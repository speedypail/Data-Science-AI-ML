```python
# Question 8
df4=pd.read_excel('probability_data.xlsx')
print(df4)
experiment = df4["Experiment"][0]
print("Experiment =", experiment)
outcomes = df4["Outcome"].unique()
print("Outcomes =", outcomes)
heads = (df4["Outcome"] == "Head").sum()
tails = (df4["Outcome"] == "Tail").sum()
total = len(df4)
probability1 = heads / total
probability2 = tails / total
print("Number of Heads =", heads)
print("Number of Tails =", tails)
print("Total Outcomes =", total)
print("Probability of Head =", probability1)
print("Probability of Tail =", probability2)
```

```python
# Question 9
import random
outcome=[]
for i in range(10):
    toss=random.choice(['Head','Tail'])
    outcome.append(toss)
print(outcome)
heads=outcome.count('Head')
tails=outcome.count('Tail')
total=len(outcome)
print("Probability of Head =", heads / total)
print("Probability of Tail =", tails / total)
```
