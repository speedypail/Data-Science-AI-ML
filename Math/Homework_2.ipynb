```python
import numpy as np
sales = [12, 15, 18, 20, 15, 22, 25, 15, 28, 30]
mean=np.mean(sales)
print(mean)
```

```python
median=np.median(sales)
print(median)
```

```python
import statistics as st
sales.sort()
mode=st.mode(sales)
print(mode)
```

```python
var=np.var(sales)
print(var)
```

```python
std=np.std(sales)
print(std)
```
