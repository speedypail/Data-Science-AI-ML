```python
import numpy as np
v1=np.array([3,4])
v2=np.array([5,7])
print("The X and Y coordinates of v1: X axis=",v1[0])
print("The X and Y coordinates of v1: Y axis=",v1[1])
```

```python
print("The X and Y coordinates of v2: X axis=",v2[0])
print("The X and Y coordinates of v2: Y axis=",v2[1])
```

```python
add=v1+v2
print('Added Vectors:',add)
```

```python
mul_a=4*v1
print("Multiplied with 4 in v1 ",mul_a)
```

```python
mul_b=2*v2
print("Multiplied with 2 in v2",mul_b)
```

```python
mag_a=np.linalg.norm(v1)
print("The magnitude of v1 is: ",mag_a)
```

```python
mag_b=np.linalg.norm(v2)
print("The magnitude of v2 is: ",mag_b)
```
