```python
def delivery_charge(distance):
    return 50+15*15
distances=[]
charges=[]
for d in range(0,6):
    charge=delivery_charge(d)
    distances.append(d)
    charges.append(charge)
    print(f"{d}\t\t₹{charge}")
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(6, 4))
plt.plot(distances, charges, marker='o', linewidth=2)
plt.title("Delivery Charge vs Distance")
plt.xlabel("Distance (km)")
plt.ylabel("Delivery Charge (₹)")
plt.grid(True)

plt.show()
```
