```python
marks=[45, 50, 55, 60, 60, 65, 70, 75, 60, 80]
mean=sum(marks)/len(marks)
print("Mean: ",mean)
```

```python
import statistics as st
marks.sort()
median=st.median(marks)
print("Median",median)
```

```python
mode=st.mode(marks)
print("Mode",mode)
```

```python
import numpy as np
varience=np.var(marks)
standard_deviation=np.std(marks)
print("Varience:",varience)
print("Standard Deviation",standard_deviation)
```

```python
difference=[(x-mean) for x in marks]
print(difference)
```
