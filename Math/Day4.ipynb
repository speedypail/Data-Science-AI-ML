```python
import numpy as np
sales_data = np.array([   [120, 150, 180],   [100, 130, 170],   [140, 160, 190]])
shape=sales_data.shape
print('The shape of the matrix is: ',shape)
```

```python
bonus_sales = np.array([ [10, 15, 20], [5, 10, 15],  [12, 14, 18]])
result=sales_data+bonus_sales
print('The result of sales and bonus is:\n', result)
```

```python
mul=3*sales_data
print(mul)
```

```python
transpose=sales_data.T
print('The transpose of the sales_data is:\n',transpose )
```

```python
A = np.array([ [4, 2, 1], [3, 5, 2],[1, 2, 3]])
determinant=np.linalg.det(A)
print('The deteminant of the matrix A is: ',determinant)
```

```python
if determinant==0:
    print('Singular')
else:
    print('Non Singular')
```

```python
inverse=np.linalg.inv(A)
print(inverse)
```

```python
identity=np.eye(3)
print(identity)
```

```python
rank=np.linalg.matrix_rank(A)
print('The rank of matrix A is:', rank)
```
