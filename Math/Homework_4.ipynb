```python
import numpy as np
student_scores = np.array([  [75, 82, 88],  [68, 79, 91],   [80, 85, 95]])
shape=student_scores.shape
print('The shape of the student_scores is: ',shape)
```

```python
improvement_scores = np.array([ [5, 4, 3],[6, 5, 4],  [4, 3, 2]])
result=student_scores+improvement_scores
print('The result of 2 data are:\n',result)
```

```python
mul=2*student_scores
print(mul)
```

```python
transpose=student_scores.T
print('The transpose of studnet scores is:\n',transpose)
```

```python
B = np.array([[2, 1, 3],[1, 4, 2], [3, 2, 5]])
det=np.linalg.det(B)
print('The determinant of B is:',det)
```

```python
if det==0:
    print('Singular')
else:
    print('Non-Singular')
```

```python
inverse=np.linalg.inv(B)
print('The inverse of matrix B is:\n',inverse)
```

```python
identity=np.eye(4)
print(identity)
```

```python
rank=np.linalg.matrix_rank(B)
print('The rank of matrix B is: ',rank)
```
