```python
def membership_const(months):
    return 500+300*months
months_list=[]
cost_list=[]
for month in range(0,7):
    cost=membership_const(month)
    months_list.append(month)
    cost_list.append(cost)
    print(f"{month}\t₹{cost}")
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(6,4))
plt.plot(months_list, cost_list, marker='o', linewidth=2)
plt.title("Months vs Membership Cost")
plt.xlabel("Months")
plt.ylabel("Membership Cost (₹)")
plt.grid(True)
plt.show()
```
