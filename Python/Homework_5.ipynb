```python
# Customer Purchase Sets
mobile_buyers={"Legolas","James bond","Ethan hunt","Walter White","Jackie Chan"}
laptop_buyers={"Mr Bean","Ethan hunt","Walter White","Jessy Pinkman","Jin Sakai"}
both=mobile_buyers.intersection(laptop_buyers)
print(f"Customers who brought mobile and laptop: {both}")
only_mobile=mobile_buyers.difference(laptop_buyers)
print(f"Customers who brought only mobile phone: {only_mobile}")
only_laptop=laptop_buyers.difference(mobile_buyers)
print(f"Customers who brought only laptop: {only_laptop}")
unique=mobile_buyers.union(laptop_buyers)
print(f"Unique customers are: {unique}")
print(f"Total unique customers is: {len(unique)}")
unique.remove("Legolas")
print(f"The updated customers are {unique}")
```

```python
# Customer Profile Dictionary
customers={
    100: {"Name": "Ethan Hunt",
          "city": "Mumbai",
          "total_purchase": 25000},
    101:  {"Name": "Walter White",
           "city": "Trivandrum",
           "total_purchase": 75000},
    102:  {"Name": "Jessy Pinkman",
           "city": "Delhi",
           "total_purchase": 50000},
}
for cid in customers:
    print(f"The customers id\'s are:{cid}")

for cid, details in customers.items():
    if details["total_purchase"] > 50000:
        print(f"{cid} -> {details['Name']} has high purchase")
customers.update({
   103: {"Name": "Jackie Chan",
        "city": "Chennai",
        "total_purchase": 90000}
})
print(f"The updated customers are: {customers}")
del customers[102]
print(f"After deleteing a customer: {customers}")
copy_customers=customers.copy()
print(f"The copied customers are: {copy_customers}")
customers.clear()
print(f"Cleared customers are: {customers}")
```
