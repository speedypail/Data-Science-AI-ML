```python
Valid_marks=0
Passed_marks=0
Failed_marks=0

try: 
    with open("marks.txt","r") as file:
        for line in file:
            values=line.split()
            print(values)
            for value in values:
                try:
                    marks=int(value)
                    if marks<0 or marks>100:
                        raise ValueError("Marks out of range")
                except ValueError:
                    print(f"{value}-> Invaid Value Skipped")
                else:
                    Valid_marks+=1
                    if marks>=40:
                        Passed_marks+=1
                    else:
                        Failed_marks+=1
except FileNotFoundError:
    print("File not found")
else:
    print("\n---Summery---")
    print("Vaild Marks",Valid_marks)
    print("Passed marks",Passed_marks)
    print("Failed Marks",Failed_marks)
finally:
    print("\nProgram Execution Completed")
```
