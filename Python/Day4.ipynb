```python
marks=[75,85,62,90,55]
print("The marks are:",marks)
print("The length of the marks is:",len(marks))
print("To acesses the first mark:",marks[0])
print("To acesses the last mark:",marks[4])
marks[2]=70
print("The updated marks is",marks)
marks[1:3]=[82,92]
print("Replaced marks from index 1:3 is:",marks)
marks.append(76)
print("The updated marks using append is:",marks)
marks.insert(1,60)
print("The inserted marks at index 1 is:",marks)
marks_2=[81,44]
marks.extend(marks_2)
print("The extended marks is:",marks)
marks.remove(55)
print("The updated marks after removing 55 is:",marks)
marks.pop(7)
print("The updated marks after removing 44 using pop is:",marks)
for score in marks:
    print(score)
passed_marks = [x for x in marks if x >= 60]
print("The passed marks is",passed_marks)
marks.sort()
print("The sorted marks are:",marks)
marks.sort(reverse=True)
print("The sorted marks in decending order is:",marks)
```
