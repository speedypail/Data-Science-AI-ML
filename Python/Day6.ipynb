```python
num_students=int(input("Enter the number of students "))
marks=[]
passed=0
failed=0
Valid_marks=0
for i in range(num_students):
    mark=int(input(f"Enter the marks {i+1}"))
    if mark==-1:
        break
        print("Stop it")
    if mark<0 or mark>100:
        print("Invalid mark")
        continue
        
    marks.append(mark)
    Valid_marks+=1
    if mark>=80:
        print("Excellent\n")
        passed+=1
    elif mark>=60:
        print("Good\n")
        passed+=1
    elif mark>=40:
        print("Pass\n")
        passed+=1
    else:
        print("Fail\n")
        failed+=1
print("Marks:", marks)
print("Total valid students:", Valid_marks)
print("Passed students:", passed)
print("Failed students:", failed)

print("\nAnalysis completed successfully")

```
