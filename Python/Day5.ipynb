```python
# Course Enrollment using Sets
python_students = {"Alice", "Bob", "Charlie", "David"}
data_students = {"Charlie", "David", "Emma", "Frank"}
print(f"The total number of students in python is: {len(python_students)}")
print(f"The total number of students in data is: {len(data_students)}")
both_courses=python_students.intersection(data_students)
print(f"Studnets enrolled in both the courses is: {both_courses}")
python_enrolled=python_students.difference(data_students)
print(f"Students enrolled in python only is: {python_enrolled}")
data_enrolled=data_students.difference(python_students)
print(f"Students enrolled in data professional trainee is: {data_enrolled}")
unique=python_students.union(data_students)
print(f"Unique students enrolled is: {unique}")
```

```python
# Student Details using Dictionary
students_info = {
    "Alice": {"age": 20, "course": "Python"},
    "Bob": {"age": 22, "course": "Data Professional Trainee"},
    "Charlie": {"age": 21, "course": "Python"}
}
for name in students_info.keys():
    print(name)
for name, details in students_info.items():
    print(f"{name} -> Age: {details['age']}, Course: {details['course']}")
students_info["Alice"]["Course"]="Data Professional Trainee"
print(f"Updated student info is:{students_info}")
students_info.pop("Bob")
print("After removing Bo")
```
