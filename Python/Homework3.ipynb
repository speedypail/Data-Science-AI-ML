```python
user_bio="""
Hi, I am Anjali.
I love Python programming.
Improving my programming skills is my goal
"""
print(f"The original bio is: {user_bio}")
clean_bio=user_bio.strip()
print(f"Removed unwanted spaces using strip: {clean_bio}")
print(f"Total number of character is: {len(clean_bio)}")
lines=clean_bio.split('\n')
print(f"The number of lines in the bio is: {len(lines)}")
for character in clean_bio:
    print(character)
if "Python" in clean_bio:
    print(f" To check whether \"Python\" is there in clean_bio: {True}")
if "Java" not in clean_bio:
    print(f" To check whether \"Java\" is there in clean_bio: {False}")
Replacing=clean_bio.replace('Anjali','Student')
print(f"Replacing Anjali with Student: {Replacing}")
print(f"Splitting the bio into individual lines to print the list: {lines}")
print(f"Concatinating the string {clean_bio}"+ "\nThank you for reading my bio")
print("This is a formatted form using format(): {}".format(clean_bio))
message="I love \"Python programing\" \nI am also a python developer"
print(f"Messages to be printed in qutoes and next sentence in new line: {message}")
```
