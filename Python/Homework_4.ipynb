```python
monthly_sales=(12000,15000,11000,18000,16000,14000)
print(f"The monthly sales is:{monthly_sales}")
print(f"The total number of months is: {len(monthly_sales)}")
print(f"The first month sales is: {monthly_sales[0]}")
print(f"The last month sales is: {monthly_sales[5]}")
if 15000 in monthly_sales:
    print(f"15000 exists in monthly sales")
else:
    print(f"15000 does not exists in monthly sales")
for sales in monthly_sales:
    print(sales)
new_monthly_sales=list(monthly_sales)
new_monthly_sales.append(17000)
new_monthly_sales.remove(11000)
monthly_sales=tuple(new_monthly_sales)
print(f"The updated monthly sales is {monthly_sales}")
a,b,*c =monthly_sales
print(f"The first monthly sales: {a}")
print(f"The second monthly sales: {b}")
print(f"The remaining monthly sales: {c}")
sorted_monthly_sales=sorted(monthly_sales)
print(f"The sorted monthly sales is: {sorted_monthly_sales}")
new_tuple=tuple(x for x in sorted_monthly_sales if x>17000)
print(f"The highest sales is: {new_tuple}")
new_tuple_1=tuple(x for x in sorted_monthly_sales if x<=12000)
print(f"The lowest sales is: {new_tuple_1}")
```
