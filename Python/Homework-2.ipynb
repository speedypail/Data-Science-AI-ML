```python
"""
This is a score and attendence details of the student 
The score is determined out of 100
"""
student_name="Vaibhav"
print("Name of the student:",student_name)
exam_score=80
attendance_percentage=100
bonus_marks=10
final_score=exam_score+bonus_marks
print("The previous score is ",final_score)
update_score=5
update_score+=final_score
print("The final score is ",update_score)
print("Is the final score >=75?:",final_score>=75)
eligiblity=(final_score>=75) and (attendance_percentage>=80)
print("The attendance percentage of",student_name,"is",attendance_percentage,"%")
print("Eligibilty status:",eligiblity)
```
