```python
def classify_sales(amount):
    if amount>=100000:
        return "High Performence"
    elif amount>=50000:
        return "average performence"
    else:
        return "Low Performence"
calculate_bonus = lambda amount: amount * 0.05
def total_sales_recursive(sales_list, n):
    if n == 0:
        return 0
    return sales_list[n-1] + total_sales_recursive(sales_list, n-1)
months = int(input("Enter number of months: "))
sales = []
print("\nEnter sales for each month:")
for i in range(months):
    amount = int(input(f"Month {i+1}: "))
    if amount < 0:
        print("Negative sales entered. Stopping input.")
        break
    if amount == 0:
        print("Zero sales entered. Skipping this month.")
        continue
    sales.append(amount)
print("\nSales Report")
for i, amount in enumerate(sales):
    classification = classify_sales(amount)
    bonus = calculate_bonus(amount)
    print(f"Month {i+1}:")
    print(f"  Sales Amount   : {amount}")
    print(f"  Classification : {classification}")
    print(f"  Bonus (5%)     : {bonus}")
    print()
total = total_sales_recursive(sales, len(sales))
print(f"Total Sales: {total}")
```
