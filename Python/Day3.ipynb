```python
feedback_message=" Hello Team, the service was Excellent! "
print(f"The original message is:{feedback_message}")
print(f"Removed spaces using strip is:{feedback_message.strip()}")
print(f"Upper case for feedback message:{feedback_message.upper()}")
print(f"Lower case for feedback message:{feedback_message.lower()}")
print(f"The len of the feedback_message:{len(feedback_message)}")
if "service" in feedback_message:
    print(f"substring found")
if "bad" not in feedback_message:
    print(f"substring not found")
print(f"The first character of feedback_message:{feedback_message[1]}")
print(f"The last character of feedback_message:{feedback_message[-2]}")
print(f"Using slicing operator to extract \"service\" from feedback_message:{feedback_message[17:24]}")
```
