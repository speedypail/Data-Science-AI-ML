```python
import seaborn as sns
import matplotlib.pyplot as plt
tips=sns.load_dataset('tips')
print(f"The first 5 rows for tips dataset\n{tips.head()}")
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip')
plt.title('Total bill Vs Tip')
plt.show()
```

```python
sns.barplot(data=tips,x='day',y='total_bill')
plt.title('Total_bill Vs Day')
plt.show()
```

```python
sns.countplot(data=tips,x='day')
plt.title('Day Vs Count')
plt.show()
```

```python
sns.boxplot(data=tips,x='day',y='total_bill')
plt.title('Day Vs Total_bill using box plot')
plt.show()
```

```python
sns.histplot(data=tips,x='total_bill')
plt.title('Count Vs Total_bill using histogram')
plt.show()
```

```python
sns.kdeplot(data=tips,x='total_bill')
plt.title('Density Vs Total_bill')
plt.show()
```
