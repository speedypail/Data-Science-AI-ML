```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
data = {
    "id": [1,2,3,4,5,6,7,8],
    "name": ["John","Emma","Raj","Sophia","Arun","Mia","Vinay","Lily"],
    "age": [28,32,26,29,35,31,27,30],
    "salary": [54000,72000,48000,61000,82000,69000,50000,58000],
    "department": ["Sales","Marketing","IT","Sales","Finance","IT","Marketing","Finance"]
}
df=pd.DataFrame(data)
df.to_csv('sample_data.csv',index=False)
print('CSV file created sucessfully')
```

```python
sns.scatterplot(data=df,x='age',y='salary',hue='department')
plt.savefig('salary_dashboard.svg')
print('salary dashboard has been saved sucessfully')
plt.show()
```

```python
corr=df.corr(numeric_only=True)
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.show()
```
