```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
print(tips.head())
print(flights.head())
sns.set_style('whitegrid')
sns.set_context('talk')
```

```python
# Plot 1
sns.countplot(data=tips,x='day')
plt.title('Number of Customers by Day')
plt.xlabel('Day')
plt.ylabel('Count')
plt.show()
```

```python
# Plot 2
sns.lineplot(data=flights,x='passengers',y='year')
sns.set_palette('Blues')
plt.show()
```

```python
# Plot 3
custom_color=["#1f77b4", "#ff7f0e", "#2ca02c"]
sns.countplot(data=tips,x='smoker',palette=custom_color)
plt.show()
```

```python
# Plot 4
sns.set_theme()
tips = sns.load_dataset("tips")
sns.scatterplot(data=tips,x="total_bill",y="tip",hue="time",palette="coolwarm")
plt.show()
```
