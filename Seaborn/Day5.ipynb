```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
data = {
    "id": [1,2,3,4,5,6,7,8],
    "name": ["John","Emma","Raj","Sophia","Arun","Mia","Vinay","Lily"],
    "age": [28,32,26,29,35,31,27,30],
    "salary": [54000,72000,48000,61000,82000,69000,50000,58000],
    "department": ["Sales","Marketing","IT","Sales","Finance","IT","Marketing","Finance"]
}
df=pd.DataFrame(data)
df.to_csv('sample_data.csv',index=False)
print('CSV file has been created sucessfully')
```

```python
df=pd.read_csv('sample_data.csv')
sns.pairplot(df)
plt.show()
```

```python
corr=df.corr(numeric_only=True)
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.savefig('employee_correlation.png',dpi=300)
print('Figure has been saved sucessfully')
plt.show()
```
