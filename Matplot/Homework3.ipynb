```python
import matplotlib.pyplot as plt
months=[1,2,3,4]
product_a=[100,150,200,250]
product_b=[120,170,210,260]
plt.plot(months,product_a,label='Product_a')
plt.plot(months,product_b,label='Product_b')
plt.title('Monthly Product Sales Comparision',fontsize=14,color='blue')
plt.xlabel('Months',color='red')
plt.ylabel('Sales',color='green')
plt.grid(True)
plt.xticks([1,2,3,4],['Jan','Feb','Mar','Apr'])
plt.ylim(0,300)
plt.legend()
plt.show()
```
