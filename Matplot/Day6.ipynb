```python
import matplotlib.pyplot as plt
weeks = [1, 2, 3, 4]
engagement = [120, 180, 260, 340]
plt.style.use('ggplot')
plt.plot(weeks,engagement,linewidth=2,alpha=0.8)
plt.title("Campaign Engagement Growth")
plt.xlabel("Week")
plt.ylabel("Engagement Score")
plt.grid(True)
plt.tight_layout()
plt.show()
```
