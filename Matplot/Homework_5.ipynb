```python
# Task 1
import pandas as pd
import matplotlib.pyplot as plt
import datetime
dates = []
for d in range(1, 6):
    dates.append(datetime.date(2025, 1, d))
values = [100, 120, 150, 180, 220]
plt.plot(dates, values, marker='o')
plt.title("Sales Trend (Jan 1 - Jan 5, 2025)")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.show()
```

```python
# Task 2
x=[1,2,3,4]
sales=[100,200,300,400]
profit=[20,50,90,150]
fig,ax1=plt.subplots()
ax2=ax1.twinx()
ax1.plot(x,sales)
ax2.plot(x,profit)
ax1.set_xlabel("Period")
ax1.set_ylabel("Sales")
ax2.plot(x, profit, marker='s')
ax2.set_ylabel("Profit")
plt.show()
```

```python
# Task 3
x=[1,2,3,4]
y=[10,100,1000,10000]
plt.plot(x,y)
plt.yscale('log')
plt.title("Exponential Growth Analysis")
plt.show()
```
