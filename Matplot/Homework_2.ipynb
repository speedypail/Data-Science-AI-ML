```python
import matplotlib.pyplot as plt
months=[1,2,3,4]
product_a=[200,250,300,350]
product_b=[180,220,260,310]
fig,axes=plt.subplots(2,1,sharex=True)
axes[0].plot(months,product_a,linestyle='-',marker='o',color='green')
axes[1].plot(months,product_b,linestyle='--',marker='s',color='red')
axes[0].set_ylabel('Sales')
axes[1].set_ylabel('Sales')
axes[1].set_xlabel('Month')
axes[0].set_title('Product A Sales')
axes[1].set_title('Product B Sales')
plt.show()
```
