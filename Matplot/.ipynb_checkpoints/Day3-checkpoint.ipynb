```python
import matplotlib.pyplot as plt
months=[1,2,3,4]
product_a=[200,250,300,350]
product_b=[180,220,260,310]
fig,axes=plt.subplots()
axes.plot(months,product_a,linestyle='-',color='green',marker='o',label='product_a')
axes.plot(months,product_b,linestyle='--',color='red',marker='s',label='product_b')
axes.legend()
plt.show()

```

```python

fig,axes=plt.subplots(2,1,sharex=True)
axes[0].plot(months,product_a,linestyle='-',color='green',marker='o')
axes[1].plot(months,product_b,linestyle='--',color='red',marker='s')
axes[1].set_xlabel('Months')
axes[0].set_ylabel('Sales')
axes[1].set_ylabel('Sales')
axes[0].set_title('Product_A')
axes[1].set_title('Product_B')
plt.show()
```
