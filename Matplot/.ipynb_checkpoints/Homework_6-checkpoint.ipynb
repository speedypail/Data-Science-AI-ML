```python
# Task 1
import matplotlib.pyplot as plt
months=[1,2,3,4]
sales_2024=[200,260,300,380]
sales_2025=[220,280,330,420]
plt.style.use('ggplot')
plt.plot(months,sales_2024,alpha=0.7,color='red')
plt.plot(months,sales_2025,alpha=0.7,color='blue')
plt.title('Sales for the year 2024 and 2025')
plt.xlabel('months')
plt.ylabel('sales')
plt.savefig('sales_report.png',dpi=200)
plt.tight_layout()
plt.show()
```

```python
# Task 2
import numpy as np
x=np.random.rand(50)
y=np.random.rand(50)
values=np.random.rand(50)
plt.scatter(x,y,c=values,cmap='viridis')
plt.colorbar()
plt.title("Sales Intensity Distribution")
plt.show()
```

```python
# Task 3
x=[1,2,3,4]
y=[10,20,30,40]
plt.figure(figsize=(8,4))
plt.plot(x,y)
plt.title("report_chart")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.savefig('report_chart.pdf',dpi=300)
plt.savefig('report_chart.svg',dpi=300)
plt.show()
```
