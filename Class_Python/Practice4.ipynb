```python
import numpy as np
arr=np.array([40,np.nan,50,np.inf])
print(np.isnan(arr))
print(np.isinf(arr))
```

```python
import numpy as np
arr = np.array([1, np.nan, 3, np.inf])
arr[np.isnan(arr)] = 0
arr[np.isinf(arr)] = 0
print(arr)
```

```python
import numpy as np
arr = np.array([1, np.nan, 5, np.inf, 7])
clean_arr = arr[~np.isnan(arr)]
clean_arr = clean_arr[~np.isinf(clean_arr)]
print(clean_arr)
```

```python
import numpy as np
arr = np.array([1, 2, 3])
copy_arr = arr.copy()
view_arr = arr.view()
view_arr[0] = 10
print(copy_arr)
print(view_arr)
```

```python
import numpy as np
arr = np.array([5, 6, 7])
v = arr.view()
v[1] = 100
print(arr)
```

```python
import numpy as np
big = np.arange(1_000_000)
v = big.view()   # uses no extra memory
c = big.copy()   # duplicates 1 million elements
print(v)
print(c)
```
