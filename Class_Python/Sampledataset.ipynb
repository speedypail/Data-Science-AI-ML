```python
import pandas as pd
import numpy as np
import random

np.random.seed(42)

names = ["Arun", "MEERA", "rahul", "Anu", "KIRAN", "Diya", "JOHN", "sneha", "VISHNU"]
genders = ["Male", "FEMALE", "female", "MALE"]
departments = ["IT", "hr", "Finance", "SALES", "Marketing"]
cities = ["Kochi", "TRIVANDRUM", "chennai", "Bangalore", "MADURAI"]

data = {
    "EmployeeID": np.arange(1001, 1501),
    "Name": np.random.choice(names, 500),
    "Gender": np.random.choice(genders, 500),
    "Department": np.random.choice(departments, 500),
    "City": np.random.choice(cities, 500),
    "Age": np.random.randint(22, 40, 500),
    "Salary": np.round(np.random.uniform(30000, 90000, 500), 2),
    "Experience": np.random.randint(1, 15, 500),
    "PerformanceScore": np.round(np.random.uniform(2.5, 5.0, 500), 1),
    "Status": np.random.choice(
        ["Active", "ACTIVE", "inactive", "Inactive"], 500
    )
}

df = pd.DataFrame(data)

# Add missing values
for col in df.columns:
    df.loc[np.random.choice(df.index, 15, replace=False), col] = np.nan

# Replace NaN with None
df = df.replace({np.nan: None})

# Add duplicate rows
duplicates = df.sample(20, random_state=1)
df = pd.concat([df, duplicates], ignore_index=True)

# Shuffle dataset
df = df.sample(frac=1).reset_index(drop=True)

# Save CSV
df.to_csv("sample_dataset.csv", index=False)

print(df.head())

print("\nDataset Shape:", df.shape)
print("CSV file saved as sample_dataset.csv")
```
