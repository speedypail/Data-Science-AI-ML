```python
import pandas as pd
dates = ['2023-01-01', '2023-01-02', '2023-01-03']
dt=pd.to_datetime(dates)
print(dt)
```

```python
dates = {
    'orderdates':pd.to_datetime(['2023-01-01', '2023-01-02', '2023-01-03'])
}
df=pd.DataFrame(dates,index=[1,2,3])
df['Year']=df['orderdates'].dt.year
df['Month']=df['orderdates'].dt.month
df['Day']=df['orderdates'].dt.day
print(df)
```

```python
import pandas as pd
sales = pd.Series([100, 120, 150, 170])
previous_sales = sales.shift(1)
print(previous_sales)
```

```python
import pandas as pd
sales = pd.Series([100, 120, 150, 170])
sales_change = sales.diff()
print(sales_change)
```

```python
import pandas as pd
data = {'value': [10, 15, 20]}
dates = pd.to_datetime(['2023-01-01', '2023-01-02', '2023-01-03'])
df = pd.DataFrame(data, index=dates)
print(df.loc['2023-01-02'])
```

```python
import pandas as pd

date_range = pd.date_range(
    start='2023-01-01',
    periods=6,
    freq='D'
)

df = pd.DataFrame(
    {'sales': [100, 200, 150, 300, 250, 400]},
    index=date_range
)

monthly = df.resample('ME').sum()

print(monthly)
```
