```python
import pandas as pd
data = {'Department': ['HR', 'HR', 'IT', 'IT', 'Finance'],
        'Salary': [40000, 45000, 60000, 62000, 50000]}
df = pd.DataFrame(data)
grouped=df.groupby('Department')
print(grouped['Salary'].mean())
```

```python
import pandas as pd
data = {
    'Team': ['A', 'A', 'B', 'B', 'A'],
    'Gender': ['M', 'F', 'M', 'F', 'M'],
    'Score': [88, 92, 85, 90, 75]
}
df = pd.DataFrame(data)
groupby=df.groupby('Team')['Score'].mean()
groupby2=df.groupby(['Team','Gender'])['Score'].mean()
print(groupby)
print(groupby2)
```

```python
import pandas as pd
data = {
    'Team': ['A', 'A', 'B', 'B', 'A'],
    'Gender': ['M', 'F', 'M', 'F', 'M'],
    'Score': [88, 92, 85, 90, 75]
}
df = pd.DataFrame(data)
def range_func(series):
    return series.max()-series.min()
group3=df.groupby('Team')['Score'].agg(range_func)
print(group3)
```

```python
import pandas as pd
df = pd.DataFrame({
    "Department": ["IT", "HR", "IT", "HR"],
    "Salary": [50000, 40000, 55000, 42000]
})
result=df.groupby('Department').agg({
    "Salary":["sum","mean"]
})
print(result)
```

```python
import pandas as pd
df = pd.DataFrame({
    "Department": ["IT", "HR", "IT", "HR"],
    "Salary": [50000, 40000, 55000, 42000]
})
result = df.groupby("Department").agg(
    Total_Salary=("Salary", "sum"),
    Average_Salary=("Salary", "mean")
)
print(result)
```

```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["Asha", "Ravi"],
    "Math": [85, 90],
    "Science": [78, 88]
})

result = pd.melt(df, id_vars="Name", var_name="Subject", value_name="Marks")
print(result)
```

```python
import pandas as pd

df = pd.DataFrame({
    "Math": [85, 90],
    "Science": [78, 88]
}, index=["Asha", "Ravi"])

result = df.stack()
print(result)
```

```python
import pandas as pd
data = {
    'Department': ['HR', 'HR', 'IT', 'IT', 'Finance'],
    'Employee': ['A', 'B', 'C', 'D', 'E'],
    'Salary': [3000, 3200, 4000, 4100, 5000]
}
df = pd.DataFrame(data)
pivot = df.pivot_table(values='Salary', index='Department', aggfunc='mean')
print(pivot)
```
