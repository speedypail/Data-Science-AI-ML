```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV
df = pd.read_csv("beginner_sales_dataset.csv")

# Create Total Amount column
df['Total_Amount'] = df['Price'] * df['Quantity']

# Convert Purchase_Date to datetime
df['Purchase_Date'] = pd.to_datetime(df['Purchase_Date'])

# -------------------------------
# 1. Total Sales by City (Bar Plot)
# -------------------------------
city_sales = df.groupby('City')['Total_Amount'].sum()

plt.figure(figsize=(8,5))
plt.bar(city_sales.index, city_sales.values)
plt.title("Total Sales by City")
plt.xlabel("City")
plt.ylabel("Sales Amount")
plt.show()

# -------------------------------
# 2. Product-wise Sales (Pie Chart)
# -------------------------------
product_sales = df.groupby('Product')['Total_Amount'].sum()

plt.figure(figsize=(7,7))
plt.pie(product_sales,
        labels=product_sales.index,
        autopct='%1.1f%%')
plt.title("Product Sales Distribution")
plt.show()

# -------------------------------
# 3. Customer Age Distribution (Histogram)
# -------------------------------
plt.figure(figsize=(8,5))
plt.hist(df['Customer_Age'], bins=10)
plt.title("Customer Age Distribution")
plt.xlabel("Age")
plt.ylabel("Number of Customers")
plt.show()

# -------------------------------
# 4. Price vs Quantity (Scatter Plot)
# -------------------------------
plt.figure(figsize=(8,5))
plt.scatter(df['Price'], df['Quantity'])
plt.title("Price vs Quantity")
plt.xlabel("Price")
plt.ylabel("Quantity")
plt.show()

# -------------------------------
# 5. Monthly Revenue Trend (Line Plot)
# -------------------------------
monthly_sales = (
    df.groupby(df['Purchase_Date'].dt.month)['Total_Amount']
      .sum()
)

plt.figure(figsize=(10,5))
plt.plot(monthly_sales.index,
         monthly_sales.values,
         marker='o')

plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.xticks(range(1,13))
plt.grid(True)
plt.show()
```
