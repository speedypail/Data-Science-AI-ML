```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y = [10, 20, 30, 40]
plt.plot(x, y)
plt.title("Sales Growth")
plt.xlabel("Months")
plt.ylabel("Revenue")
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y1 = [10, 20, 30, 40]
y2 = [15, 25, 35, 45]
plt.plot(x, y1, label="Product A")
plt.plot(x, y2, label="Product B")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y = [5, 15, 25, 35]
plt.plot(x, y)
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y = [10, 20, 30, 40]
plt.plot(x, y)
plt.xticks([1, 2, 3, 4], ["Q1", "Q2", "Q3", "Q4"])
plt.yticks([10, 20, 30, 40])
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y = [10, 20, 30, 40]
plt.plot(x, y)
plt.xlim(0, 5)
plt.ylim(0, 50)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y = [5, 10, 15, 20]
plt.plot(x, y)
plt.title("Styled Title", fontsize=14, color="blue")
plt.xlabel("X Axis", fontsize=12)
plt.ylabel("Y Axis", fontsize=12)
plt.show()
```
