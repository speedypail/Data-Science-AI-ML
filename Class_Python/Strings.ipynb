```python
# Slice between two characters
my_string = "Hello World!"
print(my_string[1:5]) # Output: ello

# Slice From the Start
print(my_string[:5]) # Output: Hello

# Slice To the End
print(my_string[6:]) # Output: World!
```

```python
my_string = "Hello World!"
print(my_string[-6:-1]) # Output: World
```

```python
my_string = "Hello World!"

# Convert to upper case
print(my_string.upper()) # Output: HELLO WORLD!

# Convert to lower case
print(my_string.lower()) # Output: hello world!

# Remove whitespace
print(my_string.strip()) # Output: Hello World!

# Replacing String
print(my_string.replace("World", "Universe")) # Output: Hello Universe!

# Splitting String
print(my_string.split(","))
```

```python
name = 'John'
age = 30
print('My name is {0}, I am {1} years old.'.format(name, age))

person = {'name': 'John', 'age': 30}
print('My name is {name}, I am {age} years old.'.format(**person))
```
