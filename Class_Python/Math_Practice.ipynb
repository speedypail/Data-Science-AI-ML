```python
def convert_to_celsius(fahrenheit):
    return (fahrenheit - 32) * 5/9

f = 68
c = convert_to_celsius(f)

print(f"{f}°F is equal to {c:.2f}°C")
```

```python
# Two points on a line
x1, y1 = 2, 4

x2, y2 = 6, 12

slope = (y2 - y1) / (x2 - x1)
print("Slope:", slope)
```
