```python
import numpy as np

A = np.array([[2, 4],
              [1, 2]])

det = np.linalg.det(A)

if det == 0:
    print("Singular Matrix")
else:
    print("Non-Singular Matrix")
```

```python
import numpy as np

A = np.array([[1, 2],
              [2, 4]])

rank = np.linalg.matrix_rank(A)
print(rank)
```
