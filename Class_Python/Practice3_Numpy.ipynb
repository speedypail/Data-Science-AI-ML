```python
import numpy as np

arr = np.array([10, 20, 30, 40])
result = np.where(arr > 25, 'High', 'Low')
print(result)
```

```python
import numpy as np

arr = np.array([10, 20, 30, 40])
conditions = [arr < 15, arr < 35]
choices = ['Low', 'Medium']
result = np.select(conditions, choices, default='High')
print(result)
```

```python
import numpy as np

arr = np.array([10, 20, 10, 30, 20])
print(np.unique(arr))
```

```python
import numpy as np

arr = np.array([0, 0, 5])
print(np.any(arr > 0))
```

```python
import numpy as np

arr = np.array([1, 2, 3])
print(np.all(arr > 0))
```

```python
# Sorting and searching
import numpy as np

arr = np.array([30, 10, 20, 10])
print(np.sort(arr))
```

```python
import numpy as np

arr = np.array([30, 10, 20])
print(np.argsort(arr))
```
