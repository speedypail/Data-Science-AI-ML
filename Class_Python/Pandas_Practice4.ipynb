```python
import pandas as pd

df = pd.DataFrame({
    "Product": ["Apple Juice", "Mango Shake", "Orange Juice", "Banana Smoothie"]
})

mask = df["Product"].str.contains("Juice")
result = df[mask]
print(result)
```
