```python
import seaborn as sns
import matplotlib.pyplot as plt
tips=sns.load_dataset('exercise')
print(tips.head())
```

```python
sns.scatterplot(data=tips,x='time',y='pulse')
plt.plot()
```

```python
sns.barplot(data=tips,x='kind',y='pulse')
plt.show()
```

```python
sns.countplot(data=tips,x='kind')
plt.show()
```

```python
sns.boxplot(data=tips,x='kind',y='pulse')
plt.show()
```

```python
sns.histplot(data=tips,x='pulse')
plt.show()
```

```python
sns.kdeplot(data=tips,x='pulse')
plt.show()
```

```python
sns.lineplot(data=tips,x='time',y='pulse')
plt.show()
```
