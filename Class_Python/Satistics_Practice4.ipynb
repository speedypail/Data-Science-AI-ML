# Sample data: marks of students
marks = [60, 65, 70, 75, 80]

average = sum(marks) / len(marks)
print(average)

```python
# H0: average marks = 70
expected_mean = 70
print(expected_mean)
```

```python
# H1: average marks is not equal to 70
alternative = "mean != 70"
print(alternative)
```

```python
p_value = 0.03

if p_value < 0.05:
    print("Reject H0")
else:
    print("Fail to reject H0")
```

```python
alpha = 0.05
print(alpha)
```

```python
# Decision process
p_value = 0.04
alpha = 0.05

if p_value < alpha:
    print("Reject H0")
else:
    print("Fail to reject H0")
```
