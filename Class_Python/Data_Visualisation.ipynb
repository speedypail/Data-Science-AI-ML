```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv('dataset_sales.csv')
df=pd.DataFrame(data)
print(df)
plt.plot(df['Month'],df['Sales'],marker='o')
plt.grid(True)
plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

```python
plt.bar(df['Month'],df['Sales'])
plt.show()
```

```python
plt.boxplot(df["Sales"])
plt.show()
```

```python
plt.pie(df['Units_Sold'],labels=df['Month'])
plt.legend()
plt.show()
```

```python
plt.scatter(df['Units_Sold'],df['Month'])
plt.show()
```
