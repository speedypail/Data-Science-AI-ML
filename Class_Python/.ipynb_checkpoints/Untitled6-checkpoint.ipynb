```python
import numpy as np
import matplotlib.pyplot as plt

X = np.array([1, 2, 3, 4, 5])
y = np.array([1, 4, 9, 16, 25])

plt.scatter(X, y)
plt.show()
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline

X = np.array([[500], [1000], [1500], [2000], [2500]])
y = np.array([20, 40, 65, 90, 130])

simple_model = LinearRegression()
simple_model.fit(X, y)

complex_model = Pipeline([
    ('poly', PolynomialFeatures(degree=4)),
    ('model', LinearRegression())
])
complex_model.fit(X, y)

print("Simple model prediction:", simple_model.predict([[1800]]))
print("Complex model prediction:", complex_model.predict([[1800]]))
```

```python
from sklearn.linear_model import Lasso
import numpy as np

X = np.array([
    [1200, 3, 1],
    [1500, 4, 1],
    [800, 2, 0],
    [2000, 5, 1]
])

y = np.array([300000, 400000, 180000, 500000])

model = Lasso(alpha=0.1)
model.fit(X, y)

print(model.coef_)
```

```python
from sklearn.linear_model import Ridge
import numpy as np

X = np.array([
    [1200, 3, 1],
    [1500, 4, 1],
    [800, 2, 0],
    [2000, 5, 1]
])

y = np.array([300000, 400000, 180000, 500000])

model = Ridge(alpha=1.0)
model.fit(X, y)
print(model.coef_)
```

```python
from sklearn.linear_model import Ridge

model = Ridge(alpha=1.0)
model.fit(X, y)
print(model.coef_)
```

```python
from sklearn.linear_model import ElasticNet

model = ElasticNet(alpha=0.1, l1_ratio=0.5)
model.fit(X, y)
print(model.coef_)
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(model, X.reshape(-1, 1), y, cv=5)
print(scores)
```
