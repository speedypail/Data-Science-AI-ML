```python
import pandas as pd
data=pd.read_excel('Employee_Performence.xlsx')
df=pd.DataFrame(data)
print(df.head())
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
print(df.head(10))
```

```python
df['Performance_Rating'] = df.groupby('Dept')['Performance_Rating'].transform(
    lambda x: x.fillna(x.mean())
)
print(df.head(10))
```

```python
df = df.drop_duplicates()
print(df.head(10))
```

```python
df['Joining_Date'] = pd.to_datetime(df['Joining_Date'])
print(df.head())
```

```python
df.rename(columns={'Dept': 'Department'}, inplace=True)
print(df.head())
```

```python
df['Annual_Salary'] = df['Monthly_Salary'] * 12
print(df.head())
```

```python
df['Attendance_Percentage'] = (df['Present_Days'] / df['Working_Days']) * 100
print(df.head())
```

```python
def performance_category(rating):
    if rating >= 4.5:
        return "Excellent"
    elif rating >= 3.5:
        return "Good"
    elif rating >= 2.5:
        return "Average"
    else:
        return "Poor"

df['Performance_Category'] = df['Performance_Rating'].apply(performance_category)
print(df.head())
```

```python
def age_group(age):
    if age < 25:
        return "Young"
    elif age <= 35:
        return "Adult"
    else:
        return "Senior"
df['Age_Group'] = df['Age'].apply(age_group)
print(df.head())
```

```python
df['Gender'] = df['Gender'].astype('category')
df['Department'] = df['Department'].astype('category')
df['City'] = df['City'].astype('category')
print(df.head())
print(df.dtypes)
```

```python
print("\nEmployees in IT Department")
print(df[df['Department'] == 'It'])

print("\nSalary > 60000")
print(df[df['Monthly_Salary'] > 60000])

print("\nAge between 25 and 35")
print(df[(df['Age'] >= 25) & (df['Age'] <= 35)])

print("\nAttendance > 90")
print(df[df['Attendance_Percentage'] > 90])

print("\nExcellent Employees")
print(df[df['Performance_Category'] == 'Excellent'])
print(df.head())
```

```python
print("\nTotal Employees by Department")
print(df.groupby('Department')['Employee_ID'].count())

print("\nAverage Salary by Department")
print(df.groupby('Department')['Monthly_Salary'].mean())

print("\nAverage Rating by Department")
print(df.groupby('Department')['Performance_Rating'].mean())

print("\nHighest Attendance")
print(df['Attendance_Percentage'].max())

print("\nLowest Attendance")
print(df['Attendance_Percentage'].min())

print("\nAverage Attendance")
print(df['Attendance_Percentage'].mean())
```
