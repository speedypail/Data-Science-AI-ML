```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("whitegrid")

tips = sns.load_dataset("tips")
sns.barplot(data=tips, x="day", y="total_bill")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_context("talk")

tips = sns.load_dataset("tips")
sns.lineplot(data=tips, x="size", y="total_bill")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_palette("pastel")

tips = sns.load_dataset("tips")
sns.boxplot(data=tips, x="day", y="total_bill")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("whitegrid")
sns.set_context("talk")
sns.set_palette("deep")

tips = sns.load_dataset("tips")
sns.scatterplot(data=tips, x="total_bill", y="tip")

plt.title("Total Bill vs Tip")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_palette("Set2")
tips = sns.load_dataset("tips")

sns.countplot(data=tips, x="day")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_palette("Blues")
flights = sns.load_dataset("flights")

sns.lineplot(data=flights, x="year", y="passengers")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme()
flights = sns.load_dataset("flights").pivot(index="month", columns="year", values="passengers")

sns.heatmap(flights, cmap="coolwarm")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

custom_colors = ["#1f77b4", "#ff7f0e", "#2ca02c"]

tips = sns.load_dataset("tips")
sns.countplot(data=tips, x="smoker", palette=custom_colors)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme()

tips = sns.load_dataset("tips")

sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="time",
    palette="coolwarm"
)
plt.show()
```
