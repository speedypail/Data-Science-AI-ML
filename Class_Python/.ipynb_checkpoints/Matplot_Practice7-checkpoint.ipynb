```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y = [10, 20, 30, 40]
plt.plot(x, y, color="green")
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1, 2, 3]
y1 = [10, 20, 30]
y2 = [15, 25, 35]
plt.plot(x, y1, color="blue")
plt.plot(x, y2, color="orange")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.random.rand(50)
y = np.random.rand(50)
values = np.random.rand(50)
plt.scatter(x, y, c=values, cmap="viridis_r")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
data = np.random.rand(10, 10)
plt.imshow(data, cmap="bone")
plt.colorbar()
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y1 = [10, 20, 30, 40]
y2 = [15, 25, 35, 45]
plt.plot(x, y1, alpha=0.5)
plt.plot(x, y2, alpha=0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.style.use("ggplot")
plt.plot([1, 2, 3], [10, 20, 30])
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y = [10, 20, 30, 40]
plt.plot(x, y)
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Simple and Clear Chart")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y = [5, 10, 15, 20]
plt.plot(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.style.use("seaborn-v0_8")
x = [1, 2, 3, 4]
y = [12, 18, 25, 35]
plt.plot(x, y, linewidth=2)
plt.title("Quarterly Performance", fontsize=14)
plt.xlabel("Quarter")
plt.ylabel("Revenue")
plt.tight_layout()
plt.show()
```
