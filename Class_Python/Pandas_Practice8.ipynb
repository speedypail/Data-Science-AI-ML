```python
import pandas as pd
data=pd.read_excel('Employee_Performence.xlsx')
df=pd.DataFrame(data)
print(df.head())
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
print(df.head(10))
```

```python
df['Performance_Rating'] = df.groupby('Dept')['Performance_Rating'].transform(
    lambda x: x.fillna(x.mean())
)
print(df.head(10))
```
