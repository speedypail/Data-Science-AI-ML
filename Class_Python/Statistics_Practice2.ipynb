```python
import random

results = []
for _ in range(100):
    results.append(random.choice(["Head", "Tail"]))
    
    heads = results.count("Head")
    
print(heads / 100)
```

```python
# Dataset example
students = [
    {"laptop": True, "python": True},
    {"laptop": True, "python": False},
    {"laptop": False, "python": False},
    {"laptop": True, "python": True}
]

B = [s for s in students if s["laptop"]]
A_and_B = [s for s in students if s["laptop"] and s["python"]]

probability = len(A_and_B) / len(B)
print(probability)
```
