```python
import matplotlib.pyplot as plt
categories = ["A", "B", "C"]
values = [30, 45, 25]
plt.bar(categories, values)
plt.show()
```

```python
import matplotlib.pyplot as plt
labels = ["Q1", "Q2", "Q3"]
sales_2024 = [20, 35, 30]
sales_2025 = [25, 30, 40]
x = range(len(labels))
plt.bar(x, sales_2024, label="2024")
plt.bar(x, sales_2025, bottom=sales_2024, label="2025")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2
plt.contour(X, Y, Z)
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
z = [5, 15, 10, 20]
ax.scatter(x, y, z)
plt.show()
```
