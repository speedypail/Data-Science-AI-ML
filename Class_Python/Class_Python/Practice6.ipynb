```python
import pandas as pd
import sqlite3

# Step 1: Connect database
conn = sqlite3.connect("company.db")

# Step 2: Create cursor
cursor = conn.cursor()

# Step 3: Create table
cursor.execute("""
CREATE TABLE employees (
    id INTEGER,
    name TEXT,
    salary INTEGER
)
""")

# Step 4: Insert records
cursor.execute("INSERT INTO employees VALUES (1, 'Arun', 30000)")
cursor.execute("INSERT INTO employees VALUES (2, 'Meera', 40000)")
cursor.execute("INSERT INTO employees VALUES (3, 'Rahul', 35000)")

# Step 5: Save changes
conn.commit()

# Step 6: Read SQL table into DataFrame
df = pd.read_sql("SELECT * FROM employees", conn)

# Step 7: Print DataFrame
print(df)

# Step 8: Close connection
conn.close()
```
