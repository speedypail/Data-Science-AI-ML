```python
# Recursion
def factorial(n):
    if n == 1:
        return 1
    else:
        return n * factorial(n - 1)

print(factorial(5))
```

```python
# Base case
def countdown(n):
    if n == 0:
        print("Done")
    else:
        print(n)
        countdown(n - 1)

countdown(3)
```

```python
# Lamda Functions
add = lambda x, y: x + y
print(add(2, 3))
```

```python
# Lambda with built in functions
numbers = [1, 2, 3, 4]
squared = list(map(lambda x: x * x, numbers))
print(squared)
```
