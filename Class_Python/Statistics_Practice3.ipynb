```python
# Email dataset
emails = [
    {"spam": True, "offer": True},
    {"spam": True, "offer": True},
    {"spam": False, "offer": True},
    {"spam": False, "offer": False}
]

# Calculate probabilities
A = [e for e in emails if e["spam"]]
B = [e for e in emails if e["offer"]]
B_given_A = [e for e in emails if e["spam"] and e["offer"]]

P_A = len(A) / len(emails)
P_B = len(B) / len(emails)
P_B_given_A = len(B_given_A) / len(A)

# Apply Bayes
P_A_given_B = (P_B_given_A * P_A) / P_B

print(P_A_given_B)
```
