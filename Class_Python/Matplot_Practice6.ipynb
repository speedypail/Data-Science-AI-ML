```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y = [10, 15, 20, 25]
plt.plot(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([1, 2, 3, 4])
y = np.array([5, 10, 15, 20])
plt.plot(x, y,linestyle='--',marker='|')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
data = pd.Series([10, 20, 30, 40],index=['a','b','c','d'])
data.plot()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = [1, 2, 3, 4]
y = [10,np.nan, 30, 40]
plt.plot(x, y,marker='o')
plt.show()
```

```python
import matplotlib.pyplot as plt
import datetime
dates = []
for d in range(1,5):
    dates.append(datetime.date(2025,1,d))
values = [100, 120, 140, 160]
plt.plot(dates, values)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y = [10, 20, 30, 40]
plt.plot(x, y)
plt.annotate("Peak", xy=(4, 40), xytext=(3, 30), arrowprops=dict(arrowstyle="->"))
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y = [10, 20, 30, 40]
plt.plot(x, y)
plt.scatter([3,4], [30,40], color="red")
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y1 = [10, 20, 30, 40]
y2 = [100, 200, 300, 400]
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.plot(x, y1,color='red',linestyle='--',marker='o')
ax2.plot(x, y2,color='green',linestyle=':',marker='s')
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y = [10, 100, 1000, 10000]
plt.plot(x, y)
plt.yscale("log")
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y1 = [10, 20, 30, 40]
y2 = [5, 10, 15, 20]
plt.stackplot(x, y1, y2)
plt.show()
```
