```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv('sales_data.csv')
df=pd.DataFrame(data)
df.head()
```

```python
plt.plot(df['Month'], df['Laptop_Sales'], marker='o', label='Laptop')
plt.plot(df['Month'], df['Mobile_Sales'], marker='s', label='Mobile')
plt.plot(df['Month'], df['Tablet_Sales'], marker='D', label='Tablet')
plt.title('Monthly Product Sales')
plt.xlabel('Month')
plt.ylabel('Unit Sold')
plt.legend()
plt.grid(True)
plt.show()
```
