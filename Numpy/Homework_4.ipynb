```python
# Task 1
import numpy as np
scores=np.array([25, 45, 65, 85, 95])
result=np.where(scores<45,"Poor","Good")
print(f"The results are: {result}")
```

```python
# Task 2
condition=(scores<40,scores<70,scores<90)
choice=("Poor","Average","Good")
result2=np.select(condition,choice,default="Excellent")
print(f"The results are: {result2}")
```
