```python
import numpy as np
salaries=np.array([35000, 42000, np.nan, 50000, np.inf, 38000, np.nan])
print(f"The salaries are {salaries}")
detect_nan=np.isnan(salaries)
print(f"The detected is_nan in salaries is: {detect_nan}")
detect_inf=np.isinf(salaries)
print(f"The detected is_inf in salaries is: {detect_inf}")
salaries[np.isinf(salaries)]=30000
salaries[np.isnan(salaries)]=30000
print(f"The cleaned valid salaries are: {salaries}")
```

```python
# Part-2
arr=np.array([35000, 42000, np.nan, 50000, np.inf, 38000, np.nan])
clean_arr=arr[~np.isnan(arr)]
clean_arr=clean_arr[~np.isinf(clean_arr)]
print(f"The cleaned array is: {clean_arr}")
```

```python
# Part-3
arr1=np.array([10,20,30,40])
copy_arr=np.copy(arr1)
print(f"The copied elements from array is: {copy_arr}")
view_arr=arr1.view()
print(f"The viewed elemets from array is: {view_arr}")
view_arr[0]=999
print(f"The modified array is: {arr1}")
```
