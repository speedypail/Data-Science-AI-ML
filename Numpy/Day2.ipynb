```python
# Task 1
import numpy as np
print(f"The 5 random numbers between 0 and 1 is:\n{np.random.rand(5)}")
print(f"The 5 random numbers between 0 and 1 using normal distribution is:\n{np.random.randn(5)}")
```

```python
# Task 2
arr=np.random.randint(35,100,(3,4))
print(f"The 3×4 array of random integers between 35 and 100\n{arr}")
```

```python
# Task 3
arr=[101, 102, 103, 104, 105]
print(f"The randomly selected students is:{np.random.choice(arr,size=2)}")
```

```python
# Task 4
np.random.seed(10)
print(f"Generating 3 random values{np.random.rand(3)}")
```

```python
# Task 5
arr=np.array([10, 20, 30, 40, 50, 60])
print(f"The 3rd element in the array is: {arr[2]}")
print(f"Elements from index 1 to 4 is: {arr[1:4]}")
print(f"Element from the last array is: {arr[-1]}")
print(f"To print every second element in an array is: {arr[::2]}")
print(f"Reversing every element in an array is: {arr[::-1]}")
```
