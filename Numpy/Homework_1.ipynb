```python
# Task 1
import numpy as np
arr1=np.arange(101,121)
print(f"The array to be printed is :\n{arr1}")
print(f"The shape of the matrix is :{arr1.shape}")
print(f"The size of the matrix is :{arr1.size}")
print(f"The data type of the matrix is : {arr1.dtype}")
```

```python
# Task 2
arr2=np.full((5,5),1)
print(f"The 5 x 5 matrix to be filled with 1 is:\n {arr2}")
print(f"The dimension of the array is: {arr2.ndim}")
```

```python
# Task 3
arr3=np.zeros((4,3))
print(f" 4×3 array of zeros to represent initial scores of 4 students in 3 subjects:\n {arr3}")
```

```python
# Task 4
arr4=np.linspace(0,100,6)
print(f"The 6 evenly spaced values between 0 and 100 is:\n {arr4}")
# Linspace can be used for scientifc computing and data visualisation
```

```python
# Task 5
arr5=np.eye(5)
print(f"The identity matrix with 5 x 5 size is:\n {arr5}")
arr6=np.diag([2,4,6,8,10])
print(f"The diagonal matrix with the number given is:\n {arr6}")
```
