```python
import numpy as np
marks=np.array([78, 85, 62, 90, 55, 88])
total_marks=np.sum(marks)
print(f"The total marks of the students is: {total_marks}")
average=np.mean(marks)
print(f"The average marks of the student is: {average}")
median=np.median(marks)
print(f"The median value of the marks is: {median}")
standard_deviation=np.std(marks)
print(f"The standard devaition of student marks is: {standard_deviation}")
min_marks=np.min(marks)
print(f"The minimum value of student marks is: {min_marks}")
max_marks=np.max(marks)
print(f"The maximum value of student marks is: {max_marks}")
grace_marks=marks+5
print(f"The grace marks for each student is {grace_marks}")
mul_marks=marks*1.1
print(f"Multiplying all marks for the student is: {mul_marks}")
```
