```python
# Part-1
import numpy as np
daily_sales=np.array([250.75, 300.40, 275.60, 400.90, 150.30, 500.50, 350.80])
print(f"The daily sales are:\n{daily_sales}")
total_weekly_sales=np.sum(daily_sales)
print(f"The daily total weekly sales is: {total_weekly_sales}")
mean_sales=np.mean(daily_sales)
print(f"The average for daily sales is: {mean_sales}")
median_sales=np.median(daily_sales)
print(f"The median sales are: {median_sales}")
variation=np.std(daily_sales)
print(f"The varaiation in daily sales is: {variation}")
minimum=np.min(daily_sales)
print(f"The minimum values of daily sales is: {minimum}")
maximum=np.max(daily_sales)
print(f"The maximum value in daily sales is: {maximum}")
```

```python
# Part-2
floor=np.floor(daily_sales)
print(f"The floor values for daily sales is: {floor}")
ceil=np.ceil(daily_sales)
print(f"The ceil values for daily sales is: {ceil}")
rounding=np.round(daily_sales)
print(f"The rounded values for daily sales is: {rounding}")
arr=np.array([1,2,3])
expo=np.exp(arr)
print(f"The exponential growth is: {expo}")
```

```python
# Part-3
cumilative_sum=np.cumsum(daily_sales)
print(f"The cumilative sum of daily sales is: {cumilative_sum}")
```

```python
# Part-4
a=np.array([[2,1],[3,4]])
b=np.array([[1,2],[5,6]])
multiplication=a@b
print(f"The product of two matrix is:\n{multiplication}")
determinant=np.linalg.det(a)
print(f"The determinant of matrix \"a\" is: {determinant}")
inverse=np.linalg.inv(a)
print(f"The inverse of matrix \"a\" is:\n{inverse}")
values,vectors=np.linalg.eig(a)
print(f"The eigen values of matrix A is:\n{values}")
print(f"The eigen vectors of matrix A is:\n{vectors}")
```
