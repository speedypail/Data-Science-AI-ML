```python
celsius=[28, 30, 32, 35, 33]
for temp in celsius:
    f=(temp*9/5) + 32
    print(temp,"°C =",f,"°F")
```

```python
import numpy as np
import sys
c=np.array([28, 30, 32, 35, 33])
f=(c*9/5)+32
print(c,"°C =",f,"°F")
```
