```python
# Task 1
import numpy as np
values=np.random.rand(10)
print(f"The randomly generated values is :\n{values}")
present_values=values>0.5
print("The present values are",values[present_values])
```

```python
# Task 2
score=np.random.randint(0,101,15)
print("The scores are",score)
greater=score>60
print("The score greater than 60 is:",score[greater])
print("The score which is divisible by 5 is",score[score%5==0])
```

```python
# Task 3
arr=np.array([100,200,300,400,500])
indices=[0,2,4]
print("The fancy element are:",arr[indices])
```

```python
# Task 4
arr=np.random.randint(1,50,(4,4))
print("The randomly generated array are:\n",arr)
print("The entire second row is:",arr[1,:])
print("The entire third column is",arr[:,2])
```

```python
# Task 5
uniform_val=np.random.uniform(0,1,size=5)
normal_val=np.random.normal(0,1,size=5)
print("The uniform values are ",uniform_val)
print("The normal values are ",normal_val)
```
