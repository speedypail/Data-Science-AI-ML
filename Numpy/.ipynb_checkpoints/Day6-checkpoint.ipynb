```python
import numpy as np
sensor=np.array([25.5, 26.0, np.nan, 27.8, np.inf, 26.4])
print(f"The sensor reading are: {sensor}")
missing=np.isnan(sensor)
print(f"Missing elements using is_nan(): {missing}")
missing1=np.isinf(sensor)
print(f"Missing elements using is_inf(): {missing1}")
sensor[np.isnan(sensor)]=0
print(f"The relaced sensor value for is_nan is:\n {sensor}")
sensor[np.isinf(sensor)]=0
print(f"The replaced sensor value for ia_inf is: \n {sensor}")
print(f"The cleaned sensor readings are: {sensor}")
```
