```python
import numpy as np
marks=np.array([45, 55, 60, 70, 85])
grace_marks=marks+5
print(f"The grace marks are: {grace_marks}")
normalise=(marks-np.min(marks))/(np.max(marks)-np.min(marks))
print(f"The normalise marks are: {normalise}")
```

```python
np.savetxt("data.txt",normalise)
```

```python
np.genfromtxt("data.txt")
```

```python
import pandas as pd
df=pd.DataFrame(normalise,columns=["normalised marks"])
print(df)
```
