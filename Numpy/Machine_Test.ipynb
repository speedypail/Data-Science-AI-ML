```python
# Question 1
import numpy as np
import pandas as pd
noise=pd.read_csv("noise_data.csv")
print(f"The noise dataset is:\n {noise}")
sales=pd.read_csv("sales_data.csv")
print(f"The sales dataset is:\n {sales}")
customer=pd.read_csv("customer_data.csv")
print(f"The customer dataset is:\n {customer}")
print(f"The shape of the sales dataset is: {sales.shape}")
print(f"The shape of the noise dataset is: {noise.shape}")
print(f"The shape of the customer dataset is: {customer.shape}")
print(f"The number of elements in sales dataset is: {sales.size}")
print(f"The number of elemnts in noise dataset is: {noise.size}")
print(f"The number of elemnts in customer dataset is : {customer.size}")
print(f"The dimension of sales dataset is: {sales.ndim}")
print(f"The dimension of noise dataset is: {noise.ndim}")
print(f"The dimension of customer dataset is: {customer.ndim}")
print(f"The data type of sales dataset is:\n{sales.dtypes}")
print(f"The data type of noise dataset is:\n{noise.dtypes}")
print(f"The data type of customer datset is:\n{customer.dtypes}")
```

```python
# Question 10
load=np.genfromtxt('sales_data.csv',delimiter=',',skip_header=1,invalid_raise=False) # invalid_raise=False to skip the inconsistent rows
print("The loaded dataset:\n",load)
average=np.nanmean(load) # nanmean is used instead of mean is because inconsistent rows get filled with NaN 
print("The average is",average)
```

```python
# Question 11
median=np.nanmedian(load)
print("The median is:",median)
standard=np.nanstd(load)
print("The standard deviation is: ",standard)
varience=np.nanvar(load)
print("The varience is: ",varience)
cum_sum=np.nancumsum(load)
print("The cum sum is: \n",cum_sum)
```

```python
# Question 12
matrix=np.array([[1,2],[3,4]])
solve=np.array([5,6])
print("The matrix is:\n",matrix)
answer=matrix@matrix
print("The product of two matrix is:\n",answer)
det=np.linalg.det(matrix)
print("The determinant of the matrix is:",det)
inv=np.linalg.inv(matrix)
print("The inverse of the matrix is:\n",inv)
linear=np.linalg.solve(matrix,solve)
print("The solved linear equation is:\n",linear)
```
