```python
# Handling missing value
df.isnull().sum()
```

```python
df['salary_in_usd'].fillna(df['salary_in_usd'].mean(), inplace=True)
print(df['salary_in_usd'])
```

```python
df['company_size'].fillna("Unknown", inplace=True)
print(df['company_size'])
```

```python
# Removing Extra spaces
df['job_title'] = df['job_title'].str.strip()
df['company_location'] = df['company_location'].str.strip()
print(df['job_title'])
print(df['company_location'])
```

```python
# Standardizing Text Case
df['job_title'] = df['job_title'].str.title()
df['job_category'] = df['job_category'].str.title()
print(df['job_title'])
print(df['job_category'])
```

```python
# Validating Experience Levels
allowed_levels = ['Entry-level','Mid-level','Senior','Executive']
df = df[df['experience_level'].isin(allowed_levels)]
print(df)
```

```python
# Removing Duplicate Records
print(df.drop_duplicates(inplace=True))
```

```python
# Aggregation and Grouped Analysis
# Job Count by Job Category
print(df.groupby('job_category').size())
```

```python
print(df.groupby('experience_level')['salary_in_usd'].mean())
```

```python
print(df.groupby('job_title')['salary_in_usd'].mean())
```

```python
print(df.groupby('experience_level')['salary_in_usd'].mean())
```

```python
print(df.groupby('work_setting')['salary_in_usd'].mean())
```

```python
print(df.groupby('job_category')['salary_in_usd'].agg(['count','mean','min','max']))
```

```python
print(df.groupby(['job_category','experience_level'])['salary_in_usd'].mean())
```

```python
print(df.groupby('company_location')['salary_in_usd'].mean().reset_index())
```

```python
print(df.groupby('job_title')['salary_in_usd'].mean().sort_values(ascending=False))
```
