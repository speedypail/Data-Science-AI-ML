```python
import pandas as pd
data=pd.read_csv('sample_dataset.csv')
df=pd.DataFrame(data)
print(df)
```

```python
df['average_income']=df['Annual Income'].mean()
print(df)
```

```python
countries=df['Country'].unique().tolist()
print(countries)
```

```python
country_count={}
for country in df['Country']:
    if country in country_count:
        country_count[country]+=1
    else:
          country_count[country]=1
print(country_count)
```

```python
for index, row in df.iterrows():
    if row['Annual Income'] > 60000:
        print(row['Name'], "- Above 60,000")
    else:
        print(row['Name'], "- Below or Equal to 60,000")
```
