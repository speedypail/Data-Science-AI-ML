```python
import pandas as pd
data = {
    "student_id": [101, 102, 103, 104, 105, 106],
    "name": ["Asha", "Rahul", "Meera", "Arjun", "Diya", "Kiran"],
    "subject": ["Maths", "Science", "English", "Maths", "Science", "English"],
    "marks": [88, 92, 76, 85, 90, 79]
}
student_df=pd.DataFrame(data)
print(student_df)
```

```python
student_df.to_csv('student.csv',index=False,encoding='utf-8')
print('student.csv file saved sucessfully')
```

```python
student_df.to_excel('student.xlsx',sheet_name='Marks_Report',index=False)
print('student.xlsx file saved sucessfully')
```

```python
student_df.to_json('student.json')
print('student.json file saved sucessfully')
```
