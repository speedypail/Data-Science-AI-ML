```python
dates2={
    'dates':pd.to_datetime(['2023-01-01', '2023-01-02' , '2023-01-03'])
}
df=pd.DataFrame(dates2)
df['Year']=df['dates'].dt.year
df['Months']=df['dates'].dt.month
df['Days']=df['dates'].dt.day
print(df)
```

```python
# Part 2
import pandas as pd
data = {'Sales': [100, 150, 120, 170, 200]}
dates = pd.to_datetime(['2024-01-01', '2024-01-02', '2024-01-03',
                        '2024-01-04', '2024-01-05'])
df = pd.DataFrame(data, index=dates)
df['Year']  = df.index.year
df['Month'] = df.index.month
df['Day']   = df.index.day
df['Previous_Day_Sales']= df['Sales'].shift(1)
df['Daily_Change']= df['Sales'].diff()
value= df.loc['2024-01-03', 'Sales']
print(df)
print(f"Sales on 2024-01-03: {value}")
```
