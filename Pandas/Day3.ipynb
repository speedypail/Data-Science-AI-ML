```python
import pandas as pd
data={
    "Name":["Asha","Ravi","Kiran","Meera","Vikram","Sona"],
    "Age":[24,27,None,26,32,25],
    "Marks":[85,90,78,88,None,75],
    "Gender":["Female","Male","Male","Female","Male","Female"],
    "Department":["HR","IT","IT","FINANCE","HR","FINANCE"]
    }
df=pd.DataFrame(data)
print(df)
first_rows=df.head()
print(f"The first 5 rows from the dataframe is :\n {first_rows}")
last_rows=df.tail()
print(f"The last 5 rows from the dataframe is :\n {last_rows}")
print(df.info())
print("The statastical information about the data is\n",df.describe())
print("Check missing values\n",df.isnull())
print("Count of missing value\n",df.isnull().sum())
df["Gender"]=df["Gender"].astype("category")
print(df.dtypes)
```
