```python
# Task 1
import pandas as pd
df=pd.read_csv("customer_data2.csv")
print("First 5 Rows:")
print(df.head())
print("\nDataset Shape:")
print(df.shape)
print("\nColumn Names:")
print(df.columns)
print("\nData Types:")
print(df.dtypes)
print("\nMissing Values:")
print(df.isnull().sum())

```

```python
# Task2
df['Age']=df['Age'].fillna(df['Age'].mean()).astype(int)
df=df.drop_duplicates()
df['Purchase_Date'] = pd.to_datetime(df['Purchase_Date'],dayfirst=True)
df = df.rename(columns={'Payment_Mode': 'PaymentType'})
print("\nData After Cleaning:")
print(df.head())
```

```python
# Task 3
df['Total_Amount']=df['Price'] * df['Quantity']
df['Age_Group']=pd.cut(
    df['Age'],
    bins=[0, 18, 35, 50, 100],
    labels=['Teen', 'Young Adult', 'Adult', 'Senior'])
df['City'] = df['City'].astype('category')
df['Product'] = df['Product'].astype('category')
df['PaymentType'] = df['PaymentType'].astype('category')

print("\nFeature Engineered Data:")
print(df.head())
print(df.dtypes)
```

```python
# Task 4
city_filter=df[df['City'] == 'Mumbai']
print("\nCustomers from Mumbai:")
print(city_filter)
age_filter=df[df['Age'] > 30]
print("\nCustomers Age > 30:")
print(age_filter)
amount_filter=df[df['Total_Amount'] > 5000]
print("\nTransactions Above 5000:")
print(amount_filter)
query_filter=df.query(
    "City == 'Mumbai' and Total_Amount > 5000"
)
print("\nQuery Filter Result:")
print(query_filter)

```

```python
# Task 5
# Total sales by city
sales_by_city=df.groupby('City')['Total_Amount'].sum()

print("\nTotal Sales by City:")
print(sales_by_city)

# Average product price
avg_price=df.groupby('Product')['Price'].mean()

print("\nAverage Product Price:")
print(avg_price)

# Total quantity sold
quantity_sold=df.groupby('Product')['Quantity'].sum()

print("\nTotal Quantity Sold:")
print(quantity_sold)
```

```python
# Task 6
# Extracting Year and Month
df['Year'] = df['Purchase_Date'].dt.year
df['Month'] = df['Purchase_Date'].dt.month
# Pivot_Table
pivot_table = pd.pivot_table(
    df,
    values='Total_Amount',
    index='City',
    columns='Product',
    aggfunc='sum',
)
print("\nPivot Table:")
print(pivot_table)
```

```python
# Crosstab Analysis
cross_tab = pd.crosstab(df['City'], df['PaymentType'])
print("\nCrosstab Analysis:")
print(cross_tab)
```

```python
# Task 7
region_data = {
    'City': ['Mumbai', 'Delhi', 'Chennai', 'Bangalore'],
    'Region': ['West', 'North', 'South', 'South']
}
region_df = pd.DataFrame(region_data)
# Mergeing with main dataset
merged_df = pd.merge(df, region_df, on='City', how='left')
print("\nMerged Dataset:")
print(merged_df.head())
```

```python
# Task 8
# Exporting the data in various format
merged_df.to_csv("Processed_Data.csv", index=False)
merged_df.to_excel("Processed_Data.xlsx", index=False)
merged_df.to_json("Processed_Data.json")

print("\nFiles Exported Successfully")
```

```python
# Task 9
aggregation = df.groupby(['City', 'Product']).agg({
    'Total_Amount': ['sum', 'mean', 'max'],
    'Quantity': ['sum', 'mean'],
    'Price': ['mean']
})
print("\nMulti-Level Aggregation:")
print(aggregation)

df['High_Value_Transaction'] = df['Total_Amount'].apply(
    lambda x: 'Yes' if x > 5000 else 'No'
)
print("\nConditional Transformation:")
print(df[['Total_Amount', 'High_Value_Transaction']].head())
```
