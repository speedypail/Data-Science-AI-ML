```python
import pandas as pd
data={
    "EmpID": [201, 202, 203, 204, 205, 206],
    "Name": ["Arun", "Neha", "Sona", "Rahul", "Maya", "John"],
    "Age": [23, 29, 35, 27, 31, 26],
    "Dept": ["IT", "HR", "IT", "Finance", "IT", "HR"],
    "Salary": [25000, 32000, 45000, 38000, 50000, 28000],
    "Attendance": [78, 88, 92, 85, 95, 60],
    "Status": ["pending", "completed", "pending", "completed", "pending", "pending"],
    "City": ["Kochi", "Trivandrum", "New Delhi", "Mumbai", "New York", "Yorkshire"]
}
df=pd.DataFrame(data)
print(df)
print("Name and Salary\n")
print("To display only Name and Salary\n",df[["Name","Salary"]])
print("To print employee details of index 3 using\n",df.loc[3])
print("Print employee details of 4th row using .iloc\n",df.iloc[3])
print(df.loc[1:4, ["Name", "Dept", "Attendance"]])
```

```python
print(df.loc[1:4, ["Name", "Dept", "Attendance"]])
```

```python
# Filtering
print(df[df["Salary"] > 35000])
```

```python
print(df[df["Salary"] > 35000])
```

```python
print(df[(df["Dept"] == "IT") & (df["Attendance"] >= 90)])
```

```python
print(df[df["City"].str.contains("York")])
```

```python
# Transformation
df["Bonus"] = df["Salary"] * 0.10
```

```python
df["FinalSalary"] = df["Salary"] + df["Bonus"]
print(df)
```

```python
df.loc[df["Status"] == "pending", "Status"] = "in progress"
print(df)
```

```python
print(df.columns)
```

```python
df = df.drop("City", axis=1, errors="ignore")
print(df)
```

```python
print(df.query('Age > 25 and Dept == "IT"'))
```

```python

df["AttendanceLevel"] = pd.cut(
    df["Attendance"],
    bins=[0, 70, 85, 100],
    labels=["Poor", "Average", "Excellent"]
)
print(df)
```

```python
df = df.applymap(lambda x: x * 1 if isinstance(x, (int, float)) else x)
print(df)
```
