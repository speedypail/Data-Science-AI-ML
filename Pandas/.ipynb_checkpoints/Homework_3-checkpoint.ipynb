```python
import pandas as pd
data={
    "EmpID":[101,102,103,104,105,106],
    "Name":["Arun","Divya","John","Meenu","Rahul","Sara"],
    "Company":["TCS","TCS","Infosys","Infosys","Wipro","Wipro"],
    "Year":[2023,2024,2023,2024,2023,2024],
    "Salary":[45000,48000,None,52000,41000,46000],
    "Role":["Developer","Tester","Developer","HR",None,"Tester"]
}
df=pd.DataFrame(data)
print(df)
first_rows=df.head(3)
print("The first 3 rows are\n ",first_rows)
last_rows=df.tail(2)
print("The last 2 rows are\n ",last_rows)
print(df.info())
print(df.describe())
print("The shape of the data is:",df.shape)
print("The columns of the data is:",df.columns)
print("The missing elements in the data is\n",df.isnull())
print("The count of missing elements in the data is\n",df.isnull().sum())
df[["Company","Role"]]=df[["Company","Role"]].astype("category")
print(df.dtypes)
```

```python
multi=[
    ["TCS","TCS","Infosys","Infosys","Wipro","Wipro"],
    ["2023","2024","2023","2024","2023","2024"]
]
index=pd.MultiIndex.from_arrays(multi,names=("Company","Year"))
df2=pd.DataFrame({"Names": ["Arun","Divya","John","Meenu","Rahul","Sara"]},index=index)
print(df2)
```
