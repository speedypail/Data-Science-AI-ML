```python
import pandas as pd
data={
    "OrderID": [1, 2, 3, 4, 5, 6],
    "CustomerName": ["Arun", "Neha", "Arun", "Sona", "Rahul", "Maya"],
    "Product": ["Apple Juice","Mango Shake","Apple Juice","Orange Juice","Banana Smoothie","Apple Juice"],
    "Qty": [2, None, 2, 1, 3, 1],
    "Price": [120, 150, 120, None, 200, 120],
    "City": ["kochi", "Delhi", "kochi", "Mumbai", "Chennai", "new york"],
    "Status": ["pending", "completed", "pending", "pending", "completed", "pending"]
}
df=pd.DataFrame(data)
print(df)
```

```python
# Part A
df['Qty']=df['Qty'].fillna(1)
df['Price']=df['Price'].fillna(0)
print(df)
```

```python
# Part B
df2=pd.DataFrame(data)
drop=df2.dropna()
print(drop)
```

```python
# Part C
duplicates=df.duplicated()
print(duplicates)
remove=df.drop_duplicates()
print(remove)
```

```python
# Part D
df['City_Upper']=df['City'].str.upper()
print(df)
```

```python
mask=df['Product'].str.contains('Juice')
result=df[mask]
print(result)
```

```python
# Part E
replace_rename=df.rename(columns={'CustomerName':'Customer'})
print(replace_rename)
rename2=replace_rename.rename(columns={'Qty':'Quantity'})
print(rename2)
convert=rename2['Quantity'].astype(int)
print(convert.dtypes)
convert2=rename2['Price'].astype(float)
print(convert2)
```

```python
# Part F
replace=replace_rename.replace({'pending':'in progress'})
print(replace)
```

```python
# Part G
sort=replace.sort_values(by='Price',ascending=False)
print(sort)
```

```python
byname=sort.set_index('Customer')
print(byname)
```

```python
reset=byname.reset_index()
print(reset)
```

```python
sort2=reset.sort_index()
print(sort2)
```
