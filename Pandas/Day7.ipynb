```python
# Step 1
import pandas as pd
data = {
    "EmpID": [101, 102, 103, 104, 105, 106],
    "Name": ["Asha", "Ravi", "Neha", "Arun", "Priya", "Kiran"],
    "Department": ["IT", "HR", "IT", "Finance", "HR", "IT"],
    "Gender": ["Female", "Male", "Female", "Male", "Female", "Male"],
    "Salary": [50000, 40000, 55000, 60000, 42000, 52000]
}
emp_df=pd.DataFrame(data)
print(emp_df)
```

```python
# Step 2
group=emp_df.groupby('Department')['Salary'].mean()
print(group)
print("\n Total aggrigation")
totalagg=emp_df.groupby('Department').agg(
    Total_Salary=('Salary','sum'),
    Average_Salary=('Salary','mean'),
    Minimum_Salary=('Salary','min'),
    Maximum_Salary=('Salary','max'),
    Emp_Count=('Name','count')
)
print(totalagg)
```

```python
# Step 3
pivot = emp_df.pivot_table(
    index='Department',
    values='Gender',
    aggfunc='count'
)

print(pivot)
```

```python
# Step 4
ct=pd.crosstab(emp_df['Gender'],emp_df['Department'])
print(ct)
```
