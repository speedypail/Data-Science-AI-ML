```python
import pandas as pd
required_columns = [
    "order_id",
    "product",
    "category",
    "price",
    "quantity"
]
dtypes = {
    "order_id": "int32",
    "product": "string",
    "category": "string",
    "price": "float32",
    "quantity": "int16"
}
filtered_chunks=[]
for chunk in pd.read_csv(
    'sales_data.csv',
    usecols=required_columns,
    dtype=dtypes,
    chunksize=1000
    
):
    filtered_chunk=chunk[chunk['quantity']>2]
    filtered_chunks.append(filtered_chunk)

final_df=pd.concat(filtered_chunks)
sample_df=final_df.sample(n=50)
sample_df.to_csv("filtered_sales.csv",index=False)
print("filtered_sales.csv saved successfully")
sample_df.to_excel("filtered_sales.xlsx",index=False)
print("filtered_sales.xlsx saved successfully")
sample_df.to_json("filtered_sales.json")
print("filtered_sales.json saved successfully")

    
```
