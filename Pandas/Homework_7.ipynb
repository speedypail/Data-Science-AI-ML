```python
# Step 1
import pandas as pd
data = {
    "OrderID": [1, 2, 3, 4, 5, 6, 7],
    "Customer": ["Anu", "Rahul", "Meera", "Ajay", "Sara", "Vishnu", "Diya"],
    "City": ["Kochi", "Kochi", "Trivandrum", "Kochi", "Trivandrum", "Kochi", "Trivandrum"],
    "Category": ["Grocery", "Electronics", "Grocery", "Clothing", "Clothing", "Grocery", "Electronics"],
    "PaymentMode": ["UPI", "Card", "Cash", "UPI", "Card", "Cash", "UPI"],
    "Amount": [1200, 25000, 800, 3500, 4000, 950, 18000]
}
orders_df=pd.DataFrame(data)
print(orders_df)
```

```python
# Step 2
group=orders_df.groupby('Category')['Amount'].sum()
print(group)
```

```python
# Step 3
multigroup=orders_df.groupby(['City','Category']).agg({
    'Amount':['mean']
})
print(multigroup)
```

```python
# Step 4
totalagg = orders_df.groupby('City').agg(
    Total_Sales=('Amount', 'sum'),
    Average_Sales=('Amount', 'mean'),
    Order_Count=('OrderID', 'count')
)

print(totalagg)
```

```python
# Step 5
pivot_table=orders_df.pivot_table(
    index='City',
    values='Amount',
    columns='PaymentMode',
    aggfunc='sum'
)
print(pivot_table)
```

```python
# Step 6
crosstab=pd.crosstab(orders_df['Category'],orders_df['PaymentMode'])
print(crosstab)
```

```python
# Step 7
pivot_reset = pivot_table.reset_index()
melted = pivot_reset.melt(
    id_vars='City',
    var_name='PaymentMode',
    value_name='TotalAmount'
)
print(melted)
```

```python
# Step 8
stacked = pivot_table.stack()
print(stacked)
```

```python
# Step 8.1
unstacked = stacked.unstack()
print(unstacked)
```
