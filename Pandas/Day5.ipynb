```python
import pandas as pd
data = {
    "StudentID": [101, 102, 103, 104, 105, 106,101],
    "Name": ["Asha", "Ravi", "Asha", "Kiran", "Meera", "John","Asha"],
    "City": ["mumbai", "Delhi", "mumbai", "chennai", "Kolkata", "Delhi","mumbai"],
    "Marks": [85, None, 85, 78, 90, 70,85],
    "Grade": ["A", "B", "A", "F", "A", "B","A"],
    "Status": ["pending", "pending", "pending", "completed", "completed", "pending","pending"]
}
df=pd.DataFrame(data)
print(df)
```

```python
# Part A
none_values=df.isnull()
print(none_values)
df_fill=df.fillna(0)
print(df_fill)
```

```python
# Part B
drop=df.dropna()
print(f"The dropped values are:\n {drop}")
```

```python
# Part C
replace=drop.replace({'F':'Fail'})
print(f"The replaced value is:\n{replace}")
```

```python
# Part D
duplicates=replace.duplicated()
print(f"The duplicates values are:\n {duplicates}")
```

```python
# Part D(A)
remove=duplicates.drop_duplicates()
print(f"Removed duplicates{remove}")
```

```python
# Part E
replace['City_Upper']=replace['City'].str.upper()
print(replace)
```

```python
# Part F
mask=replace['City'].str.contains("Del")
result=replace[mask]
print(result)
```

```python
# Part G
replace_rename=replace.rename(columns={'Marks':'Score'})
print(replace_rename)
replace_rename['Score']=replace_rename['Score'].astype(float)
print(replace_rename.dtypes)
```

```python
# Part H
sorted_df=replace_rename.sort_values(by='Score',ascending=False)
print(sorted_df)
```

```python
# Part I
sort_index=sorted_df.sort_index()
print(sort_index)
```

```python
# Part I(A)
reset_index=sort_index.reset_index(drop=True)
print(reset_index)
```

```python
final=pd.DataFrame(reset_index)
print(final)
```
