```python
import pandas as pd
emp_data = {
    "EmpID": [101, 102, 103, 104],
    "Name": ["asha", "ravi", "kiran", "meera"],
    "Dept": ["IT", "HR", "IT", "Finance"]
}

salary_data = {
    "EmpID": [101, 102, 103, 105],
    "Salary": [50000, 40000, 55000, 45000]
}

bonus_data = {
    "EmpID": [101, 102, 104],
    "Bonus": [5000, 3000, 4000]
}
emp_df=pd.DataFrame(emp_data)
salary_df=pd.DataFrame(salary_data)
bonus_df=pd.DataFrame(bonus_data)
print(emp_df)
print(salary_df)
print(bonus_df)
```

```python
# Part A
merged_df = pd.merge(emp_df, salary_df, on="EmpID", how="left")
print(merged_df)
```

```python
merged_df = pd.merge(merged_df, bonus_df, on="EmpID", how="outer")
print(merged_df)
```

```python
# Part B
merged_df["Salary"] = merged_df["Salary"].fillna(0)
merged_df["Bonus"] = merged_df["Bonus"].fillna(0)
print(merged_df)
```

```python
# Part C
merged_df["FinalSalary"] = merged_df["Salary"] + merged_df["Bonus"]
print(merged_df)
```

```python
# Part D
print(merged_df["Salary"].sum())
print(merged_df["Salary"].mean())
print(merged_df["Salary"].mean())
print(merged_df["Name"].count())
```

```python
# Part E
merged_df["Category"] = merged_df["FinalSalary"].apply(lambda x: "High" if x >= 55000 else "Medium" if 45000 <= x <= 54999 else "Low")
print(merged_df)
```

```python
# Part F
dept_dict = {
    "IT": "Information Technology",
    "HR": "Human Resources",
    "Finance": "Finance Department"
}
merged_df["Dept_FullForm"] = merged_df["Dept"].map(dept_dict)
print(merged_df)
```

```python
# Part G
merged_df["Dept_Avg_FinalSalary"] = (merged_df.groupby("Dept")["FinalSalary"].transform("mean"))
print(merged_df)
```

```python
# Part H
emp_index = emp_df.set_index("EmpID")
salary_index = salary_df.set_index("EmpID")
joined_df = emp_index.join(salary_index)
print(joined_df)
```
