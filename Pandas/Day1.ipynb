```python
import pandas as pd
a=[250, 300, 150, 400, 350]
s=pd.Series(a,index=["Mon", "Tue", "Wed", "Thu", "Fri"])
print("The series are\n",s)
sales=s['Wed']
print("The sales on Wednesday is: ",sales)
total=s.sum()
print("The total sum of the series is: ",total)
average=s.mean()
print("The average of the series is:",average)
```

```python
# Dataframe
sales_df={'Day':['Mon','Tue','Wed','Thur','Fri'],
        'Sales':[250,300,150,400,350]         
         }
df=pd.DataFrame(sales_df)
print(df)
maximum=df.max()
print(f"The maximum sales is: \n{maximum}")
minimum=df.min()
print(f"The minimum sales is:\n {minimum}")
```
