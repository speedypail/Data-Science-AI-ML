```python
import pandas as pd
data1={
    "StudentID":[1,2,3,4],
    "Name":['alice','bob','charlie','diya'],
    "Dept":['IT','HR','IT','Finance']
}
student_df=pd.DataFrame(data1)
print(student_df)
```

```python
data2={
    "StudentID":[1,2,3,5],
    "Maths":[80,45,90,70],
    "Science":[75,60,88,65]
}
marks_df=pd.DataFrame(data2)
print(marks_df)
```

```python
# Part A
merge=pd.merge(student_df,marks_df, on='StudentID',how='inner')
print(merge)
```

```python
merge2=pd.merge(student_df,marks_df, on='StudentID',how='left')
print(merge2)
```

```python
# Part B
marks_df['Total']=marks_df['Maths']+marks_df['Science'] # Vectorized operation
print(marks_df)
```

```python
# Part C
total=marks_df.sum()
print(total)
```

```python
mean=marks_df.mean()
print(mean)
```

```python
minimum=marks_df.min()
print(minimum)
```

```python
maximum=marks_df.max()
print(maximum)
```

```python
# Part D
marks_df['Result']=marks_df['Total'].apply(lambda x: 'Pass' if x>=150 else 'Fail')
print(marks_df)
```

```python
# Part E
marks_df['Percentage']=marks_df.apply(lambda row: (row['Total']/200)*100,axis=1)
print(marks_df)
```

```python
# Part F
capitalized=student_df['Name'].map(str.title)
print(capitalized)
```

```python
# Part G
final_df = pd.merge(student_df, marks_df, on='StudentID', how='left')
final_df['Dept_Avg_Total'] = final_df.groupby('Dept')['Total'].transform('mean')
print(final_df)
```

```python
# Part H
week1_df = pd.DataFrame({"StudentID": [1, 2],"Marks": [78, 82]},index=[1,2])
week2_df = pd.DataFrame({"StudentID": [3, 4],"Marks": [91, 76]},index=[1,2])
combined_df = pd.concat([week1_df, week2_df],axis=0)
print(combined_df)
```
