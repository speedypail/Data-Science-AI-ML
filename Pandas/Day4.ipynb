```python
import pandas as pd
data={
    "Student ID":[101,102,103,104,105],
    "Name":["Asha","Ravi","Kiran","Meera","John"],
    "Age":[24,30,28,26,22],
    "Dept":["HR","IT","IT","Finance","IT"],
    "Marks":[78,90,84,92,65],
    "Status":["Pending","Completed","Pending","Completed","Pending"],
    "City":["New Delhi","Mumbai","New York","New Delhi","Yorkshire"]
    
}
df=pd.DataFrame(data)
print(df)
print("Using Brackets only []\n")
print("Display only names\n",df['Name'])
print("Display name and matrks\n",df[["Marks","Name"]])
print("Using loc functions\n")
print("To print only rows using loc functions\n",df.loc[0])
print("To print marks of index 1\n",df.loc[1,"Marks"])
print("To print all rows but only names and marks\n",df.loc[:,["Name","Dept"]])
```

```python
print("To print using iloc")
print("To print the third row\n",df.iloc[2])
print("To print the value at 2nd row,5th column:",df.iloc[1,4])
print("To print all rows but only Name and Mark:\n",df.iloc[:,[1,4]])
```

```python
print("Using Substring\n")
print("Select rows 0 to 2 and columns Name and Marks\n",df.loc[0:2,["Name","Marks"]])
```

```python
print("Using Slicing\n")
print("Slicing first 3 rows using .iloc\n",df.iloc[0:3])
```

```python
print("Students with Marks >= 85: ",df[df['Marks']>=85])
result=df[(df['Age']>25) &(df['Dept']=='IT')]
print("Students whose Dept is IT AND Age > 25\n",result)
```

```python
print("Using Query\n")
result2=df.query("Marks >= 80 and Dept == 'IT'")
print(result2)
```

```python
# Transformation
df['Bonus Marks']=df['Marks']+5
print("Added Bonus Mark table\n",df)
```

```python
df.loc[df["Status"] == "Pending", "Status"] = "in progress"
print("Changed from pending to in progress\n",df)
```

```python
drop=df.drop(columns=['City'])
print(drop)
```

```python
df = df.applymap(lambda x: x * 1)
print(df)
```

```python
filtered = df[df["City"].str.contains("York")]
print(filtered)
```

```python
df["Performence"]=pd.cut(df['Marks'],bins=[0,70,85,100],labels=['Low','Average','High'])
print(df)
```
