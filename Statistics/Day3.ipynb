```python
customers = [

   {"premium": True, "promo_email": True},

   {"premium": True, "promo_email": True},

   {"premium": True, "promo_email": False},

   {"premium": False, "promo_email": True},

   {"premium": False, "promo_email": False},

   {"premium": False, "promo_email": True}

]
population = 5000
print('Population:',population)
```

```python
sample_size=len(customers)
print('The sample size of the customers dataset is:',sample_size)
```

```python
A=[c for c in customers if c['premium']]
B=[c for c in customers if c['promo_email']]
B_given_A=[c for c in customers if c['premium'] and c['promo_email']]
P_A=len(A)/len(customers)
P_B=len(B)/len(customers)
P_B_given_A=len(B_given_A)/len(A)
# Applying Bayes Theorm
P_A_given_B=(P_B_given_A*P_A)/P_B
print(f'The Bayes therom for the datatset is: {P_A_given_B}')
```
