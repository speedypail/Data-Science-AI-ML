```python
import pandas as pd
marks=[72, 78, 75, 80, None, 77, 150]
df=pd.DataFrame(marks,columns=['Marks'])
print(df)
```

```python
missing_value=df.isnull()
print(missing_value)
```

```python
print(f"Maximum mark: {df['Marks'].max()}")
```

```python
clean_marks = [72, 78, 75, 80, 77]
average_marks=sum(clean_marks)/len(clean_marks)
print('The average marks is:',average_marks)
```

```python
# H0: average marks 
expected_mean=76.4
print(expected_mean)
```

```python
# H1: average marks is not equal
alternative='mean!=76.4'
print(alternative)
```

```python
p_value=0.03
alpha=0.05
if p_value<alpha:
    print('Reject H0')
else:
    print('Failed to reject H0')
```
