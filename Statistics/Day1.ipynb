```python
student_counts = [25, 30, 28, 32, 27]
temperatures = [36.5, 37.2, 36.8, 37.0, 36.9]
favorite_subjects = ["Math", "Science", "Math", "English", "Science"]
performance_levels = ["High", "Medium", "High", "Low", "Medium"]
print('This is a discrete data:', student_counts)
print('This is a continuous data:', temperatures)
print('This is a nominal data:', favorite_subjects)
print('This is a ordinal data:', performance_levels)
```

```python
average=sum(student_counts)/len(student_counts)
print('The average of student counts is:',average)
```

```python
average_temp=sum(temperatures)/len(temperatures)
print('The average temperature is:',average_temp)
```

```python
maximum=max(student_counts)
print('The maximum student counts are:',maximum)
```

```python
minimum=min(temperatures)
print('The minimum temperature is:',minimum)
```

```python
ages = [18, 19, None, 20, 18]
for age in ages:
    if age is None:
        print('Missing value')
        break
```

```python
order={'Low':1,'Medium':2,'High':3}
numeric=[order[p] for p in performance_levels]
print('Numeric Values:',numeric)
```
