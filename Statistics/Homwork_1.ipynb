```python
projects_completed = [5, 8, 6, 10, 7]
salaries = [35000.5, 42000.0, 39000.8, 48000.2, 41000.6]
departments = ["HR", "IT", "Finance", "IT", "HR"]
performance_ratings = ["Excellent", "Good", "Excellent", "Average", "Good"]
employee_ages = [25, None, 30, 28, 26]
print('Discrete Data:',projects_completed)
print('Continuous Data:',salaries)
print('Nominal Data:',departments)
print('Ordinal Data:',performance_ratings)
```

```python
average=sum(projects_completed)/len(projects_completed)
print('The average of project completed is:',average)
```

```python
average_salaries=sum(salaries)/len(salaries)
print('The average salaries is:',average_salaries)
```

```python
maximum=max(salaries)
print('The maximum salary is:',maximum)
```

```python
minimum=min(salaries)
print('The minimum salary is:',minimum)
```

```python
if None in employee_ages:
    print('Missing Value')
else:
    print('No missing value')
```

```python
unique=set(departments)
print('Unique Departments:',unique)
```

```python
rating_order={'Average': 1,'Good':2,'Excellent':3}
ratings=[rating_order[p] for p in performance_ratings]
print('Numeric Values:',ratings)
```
