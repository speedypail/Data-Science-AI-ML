```python
import numpy as np
data=[10,20,30,40,100]
maximum=max(data)
print('The maximum value inside the data is:',maximum)
print('The value inside the data has an outlier because it has a large datapoints')
```

```python
from scipy.stats import skew
skew_data=skew(data)
print('Skewness:',skew_data)
print('The given data is postively skewed because tail is extended on the right side')
```

```python
even=[x for x in data if x%2==0]
print('The even number in the dataset is:',even)
```
