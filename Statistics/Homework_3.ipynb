```python
users = [

   {"premium": True, "new_release": True}, 

   {"premium": True, "new_release": False},

   {"premium": True, "new_release": True},

   {"premium": False, "new_release": True},

   {"premium": False, "new_release": False},

   {"premium": False, "new_release": True},

   {"premium": False, "new_release": False},

   {"premium": True, "new_release": True}

]
population=20000
print('The size of population is:',population)
```

```python
size_user=len(users)
print('The size of the sample is:',size_user)
```

```python
A=[u for u in users if u['premium']]
B=[u for u in users if u['new_release']]
B_given_A=[u for u in users if u['premium'] and u['new_release']]
P_A=len(A)/len(users)
P_B=len(B)/len(users)
P_B_given_A=len(B_given_A)/len(A)
# Applying Bayes Theorm
P_A_given_B=(P_B_given_A*P_A)/P_B
print(f'The Bayes therom for the datatset is: {P_A_given_B}')
```
