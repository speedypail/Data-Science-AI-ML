```python
import matplotlib.pyplot as plt
marks = [45, 50, 50, 60, 60, 60, 70, 80, 90, 90]
plt.hist(marks,bins=5)
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.title('Marks Distribution of the student')
plt.show()
```
