```python
import pandas as pd
data={
    'Machine Age':[1,2,3,4,5],
    'Maintenance Cost':[5,10,17,26,37]
}
df=pd.DataFrame(data)
print(df)
```

```python
import matplotlib.pyplot as plt
plt.plot(df['Machine Age'],df['Maintenance Cost'],marker='o')
plt.xlabel('Machine Age')
plt.ylabel('Maintenance Cost')
plt.title('Machine Maintenance Cost')
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
X=df[['Machine Age']]
y=df['Maintenance Cost']
poly=PolynomialFeatures(degree=2)
X_poly=poly.fit_transform(X)
print(X_poly)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_poly,y)
```

```python
import numpy as np
new_data=np.array([[6]])
new_data_poly=poly.transform(new_data)
prediction=model.predict(new_data_poly)
print(prediction)
```

```python
# Another model with degree 10
from sklearn.preprocessing import PolynomialFeatures
X=df[['Machine Age']]
y=df['Maintenance Cost']
poly=PolynomialFeatures(degree=10)
X_poly=poly.fit_transform(X)
print(X_poly)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_poly,y)
```

```python
import numpy as np
new_data=np.array([[6]])
new_data_poly=poly.transform(new_data)
prediction=model.predict(new_data_poly)
print(prediction)
```
