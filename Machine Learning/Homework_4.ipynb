```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
X = np.array([[1],
              [2],
              [3],
              [4],
              [5],
              [6],
              [7]])

y = np.array([28000, 34000, 40000, 46000, 52000, 58000, 64000])
```

```python
print("This is a Supervised Learning problem because the dataset contains both input (Years of Experience) and output (Monthly Salary). The model learns the relationship between them.")
```

```python
print("\nThis is a Regression problem because the target variable (Monthly Salary) is a continuous numerical value.")
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
prediction = model.predict([[8]])
print("\nPredicted Salary for 8 Years of Experience:")
print(prediction)
```

```python
y_pred = model.predict(X)
```

```python
mse = mean_squared_error(y, y_pred)
print("\nMean Squared Error (MSE):")
print(mse)
```

```python
print("\nModel Coefficient (Slope):")
print(model.coef_)

print("\nModel Intercept:")
print(model.intercept_)
```

```python
plt.scatter(X, y, color="blue", label="Actual Data")
plt.plot(X, y_pred, color="red", label="Regression Line")

plt.title("Years of Experience vs Monthly Salary")
plt.xlabel("Years of Experience")
plt.ylabel("Monthly Salary (₹)")
plt.legend()

plt.show()
```
