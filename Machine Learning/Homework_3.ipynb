```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression, Lasso, Ridge, ElasticNet
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline
from sklearn.model_selection import cross_val_score
```

```python
data = {
    "Age": [25, 30, 35, 40, 45, 50, 55, 60],
    "BMI": [22, 25, 27, 30, 32, 35, 37, 40],
    "Exercise Hours": [5, 4, 3, 2, 2, 1, 1, 0],
    "Medical Expense (₹ Thousands)": [90, 120, 160, 220, 280, 340, 410, 500]
}
df = pd.DataFrame(data)
print(df)
```

```python
X = df[["Age", "BMI", "Exercise Hours"]]
y = df["Medical Expense (₹ Thousands)"]
print(X)
```

```python
print(y)
```

```python
linear_model = LinearRegression()
linear_model.fit(X, y)
```

```python
poly_pipeline = Pipeline([
    ("poly", PolynomialFeatures(degree=4)),
    ("linear", LinearRegression())
])

poly_pipeline.fit(X, y)
```

```python
new_patient = [[42, 31, 2]]

linear_prediction = linear_model.predict(new_patient)
poly_prediction = poly_pipeline.predict(new_patient)

print("Linear Regression Prediction:")
print(linear_prediction)

print("\nPolynomial Regression Prediction:")
print(poly_prediction)
```

```python
lasso = Lasso(alpha=0.2)
lasso.fit(X, y)
```

```python
ridge = Ridge(alpha=2.0)
ridge.fit(X, y)
```

```python
elastic = ElasticNet(alpha=0.2, l1_ratio=0.6)
elastic.fit(X, y)
```

```python
# -------------------------
print("\nLasso Coefficients:")
print(lasso.coef_)

print("\nRidge Coefficients:")
print(ridge.coef_)

print("\nElasticNet Coefficients:")
print(elastic.coef_)
```

```python
scores = cross_val_score(elastic, X, y, cv=4)

print("\nCross Validation Scores:")
print(scores)

print("\nAverage Cross Validation Score:")
print(scores.mean())
```
