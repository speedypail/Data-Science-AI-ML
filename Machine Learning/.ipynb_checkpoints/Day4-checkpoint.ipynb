```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
X = np.array([[600],
              [900],
              [1200],
              [1500],
              [1800],
              [2100]])

y = np.array([40, 58, 76, 94, 112, 130])
```

```python
print("This is a Supervised Learning problem because")
print("the dataset contains both input (House Size) and output (Selling Price).")
print("The model learns the relationship between them.")
```

```python
print("\nThis is a Regression problem because")
print("the target variable (Selling Price) is a continuous numerical value.")
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
new_house = np.array([[1700]])
prediction = model.predict(new_house)
print("\nPredicted Selling Price for 1700 sq.ft:")
print(prediction)
```

```python
y_pred = model.predict(X)
```

```python
mse = mean_squared_error(y, y_pred)

print("\nMean Squared Error (MSE):")
print(mse)
```

```python
print("\nSlope (Coefficient):")
print(model.coef_)

print("\nIntercept:")
print(model.intercept_)
```

```python
plt.scatter(X, y, color="blue", label="Original Data")

plt.plot(X, y_pred, color="red", label="Regression Line")

plt.title("House Size vs Selling Price")
plt.xlabel("House Size (sq.ft)")
plt.ylabel("Selling Price (₹ Lakhs)")
plt.legend()

plt.show()
```
