```python
customers = [
    {"premium": True, "promo_email": True},
    {"premium": True, "promo_email": True},
    {"premium": True, "promo_email": False},
    {"premium": False, "promo_email": True},
    {"premium": False, "promo_email": False},
    {"premium": False, "promo_email": True}
]
population=5000
sample=len(customers)
premium=sum(c["premium"] for c in customers)
promo=sum(c["promo_email"] for c in customers)
premium_and_promo=sum(c["premium"] and c["promo_email"] for c in customers)
P_A=premium/sample
P_B=promo/sample
P_B_given_A=premium_and_promo/premium
P_A_given_B=(P_B_given_A * P_A)/P_B
print('Population Size :', population)
print('Sample Size:', sample)
print('P(A)=',P_A)
print('P(B)=',P_B)
print('P(B|A)=' ,P_B_given_A)
print('P(A|B)=',P_A_given_B)
```

```python
marks_quality=[72,78,75,80,None,77,150]
clean_marks=[72,78,75,80,77]
print('Missing Values')
print([x for x in marks_quality if x is None])
print('Outlier')
print([x for x in marks_quality if x==150])
average=sum(clean_marks)/len(clean_marks)
print('Average =', average)
print('Hypothesis')
print('H0 : Average marks=75')
print('H1 : Average marks=!75')
```

```python
p_value = 0.03
alpha = 0.05
if p_value < alpha:
    print('Decision : Reject H0')
    print('Conclusion : The average marks are significantly different from 75')
else:
    print('Decision : Fail to Reject H0')
    print('Conclusion : No significant evidence against H0')
```
