```python
import csv

Heading="Python Data Analytics Lab Test\n"
print(Heading.center(120))

records=[]

try:
    with open("orders.csv","r") as file:
        reader=csv.reader(file)
        header=next(reader)
        header.append("total_amount")
        print("Header:",header)

        for row in reader:
            try:
                
                if row[1]=="":
                    print("Missing customer",row)

                if row[7]=="":
                    row[7]="Pending"

                row[2]=row[2].lower()

              
                price=float(row[5])
                quantity=int(row[6])

                total_amount=price*quantity

                row.append(total_amount)
                records.append(row)

            except ValueError:
                print("Invalid numeric data:", row)

    print("\nRecords")
    for r in records:
        print(r)

except FileNotFoundError:
    print("File not found")

else:
    print("\nFile processed successfully")

finally:
    print("\nProgram execution completed")



print("\nString processing")

categories=set()
for r in records:
    categories.add(r[4])
print("Product Categories:", categories)

print("\nProduct containing 'top'")
for r in records:
    if "top" in r[3].lower():
        print(r[3])

print("\nCustomer name in upper case")
for r in records:
    r[1]=r[1].upper()
    print(r[1])

print("\nOrder Years")
for r in records:
    year=r[8].split("-")[2]
    print(year)



print("\nConditional Analysis")

print("\nHigh order price")
for r in records:
    if float(r[5])>30000:
        print(r)

print("\nCancelled Orders:")
for r in records:
    if r[7].lower()=="cancelled":
        print(r)

print("\nOrder Value Categories:")
for r in records:
    total=r[9]
    if total>50000:
        category="High Value"
    elif total>20000:
        category="Medium Value"
    else:
        category="Low Value"

    print(r[0],"->",category)


print("\nList Processing")

products=[]
for r in records:
    products.append(r[3])
print("Product Names:",products)

cities=[]
for r in records:
    if r[2] not in cities:
        cities.append(r[2])
print("Unique Cities:",cities)

print("Total Orders:",len(records))

print("\nOrders sorted by price:")
sorted_orders=records.copy()
for r in sorted_orders:
    r.insert(0,r[5])
sorted_orders.sort()
for r in sorted_orders:
    r.pop(0)

for r in sorted_orders:
    print(r)



print("\nDictionary Analysis")

city_count={}
category_count={}
category_revenue={}

for r in records:
    city=r[2]
    category=r[4]
    total=r[9]

    if city in city_count:
        city_count[city]+=1
    else:
        city_count[city]=1

    if category in category_count:
        category_count[category]+=1
    else:
        category_count[category]=1

    if category in category_revenue:
        category_revenue[category]+=total
    else:
        category_revenue[category]=total

print("\nOrder Count per City:",city_count)
print("\nOrder Count per Category:",category_count)
print("\nTotal Revenue per Category:",category_revenue)



def calculate_total(price, quantity):
    return price*quantity

def get_high_value_orders(orders):
    result=[]
    for r in orders:
        if r[9]>30000:
            result.append(r)
    return result

def orders_by_city(orders, city):
    result=[]
    for r in orders:
        if r[2]==city:
            result.append(r)
    return result

print("\nFunctions Output")

print("\nCalculate Total ")
print(calculate_total(200,3))

print("\nHigh Value Orders:")
for r in get_high_value_orders(records):
    print(r)

print("\nOrders from delhi:")
for r in orders_by_city(records,"delhi"):
    print(r)



print("\nLambda Functions")

print("\nSorted by Total Amount:")
for r in sorted(records, key=lambda x:x[9]):
    print(r)

print("\nPrice Values:")
print(list(map(lambda x:x[5], records)))

print("\nMaximum Order Value:")
print(max(records, key=lambda x:x[9]))
print("\nExporting to clean_orders.csv")

with open("clean_orders.csv", "w") as file:
    writer = csv.writer(file)
    writer.writerow(header)
    writer.writerows(records)

print("Data exported successfully")
```
